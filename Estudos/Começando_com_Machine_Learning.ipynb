## **História da Inteligência Artificial (IA)**

As discussões sobre inteligência artificial começaram a surgir graças ao trabalho de Alan Turing, matemático e cientista da computação que trouxe a partir da sua publicação de *Computing Machinery and Intelligence (TURING, 1950)* a pergunta "As máquinas podem pensar?". Neste mesmo artigo o Turing propôs uma maneira de identificar se a máquina era capaz de pensar ou não, este teste ficou conhecido como:

### Teste de Turing

Com uma proposta simples para a época, entre duas cabines teriamos um humano e um computador do qual uma terceira pessoa não saberia a priori quem reside dentro destas cabines. A ideia era que o entrevistador fizesse perguntas escritas tanto para a máquina quanto para o humano, ao fim se o entrevistador não conseguisse identificar quem era humano ou máquina, seria provada a inteligência da máquina.

---

### **Após Turing**

Apesar da repercussão teórica gerada por Turing, o termo "Inteligência Artificial" foi cunhado oficialmente apenas em 1956, durante a histórica conferência de Dartmouth, organizada por John McCarthy. McCarthy também foi fundamental ao criar a linguagem LISP (1958), que se tornou o padrão para programação simbólica de IA por décadas. Paralelamente, Arthur Samuel popularizou o termo "Machine Learning" em 1959, demonstrando a viabilidade do aprendizado de máquina através de um programa capaz de aprender a jogar damas e superar amadores.
Um marco crucial de hardware e software ocorreu em 1958, quando Frank Rosenblatt desenvolveu o Perceptron Mark I, o primeiro neurocomputador baseado em redes neurais biológicas. Diferente da computação convencional, o Perceptron demonstrava capacidade de aprendizado por ajuste de pesos sinápticos.
No entanto, a área enfrentou severos retrocessos conhecidos como "Invernos da IA". O primeiro grande inverno foi precipitado em 1969, quando Marvin Minsky e Seymour Papert publicaram o livro Perceptrons. Eles provaram matematicamente que o modelo de Rosenblatt (de camada única) era incapaz de resolver problemas não-lineares simples, como a porta lógica XOR. Essa limitação teórica, somada à escassez de poder computacional, congelou o financiamento de pesquisas por anos.
O renascimento da área ocorreu na década de 1980 com a popularização do algoritmo de Backpropagation (Rumelhart, Hinton e Williams, 1986), que permitiu o treinamento eficiente de redes neurais multicamadas, superando as limitações apontadas por Minsky. Desde os anos 2000, a convergência entre o aumento massivo de dados (Big Data) e o processamento paralelo em GPUs permitiu o surgimento do Deep Learning, levando a IA ao estado da arte atual.

## **O que é o aprendizado de máquina?**

O aprendizado de máquina faz parte do escopo da inteligência artificial, e ele tem a capacidade de criar modelos capazes de identificar padrões para realizar previsões sobre o futuro daquele conjunto de dados ou até mesmo para classificar uma determinada informação dividindo um dado entre grupo X e Y com base em características pré-disponibilizadas. Entretanto, apesar disto tudo o aprendizado de máquina não se prende a apenas estes dois, se subdividindo em alguns tipos de aprendizado que iremos abordar mais a frente para que todos tenham uma compreensão sólida sobre quais são suas especificidades.

## **Fundamentos de Dados (Features, Labels e Datasets)**

Quando vamos trabalhar com Machine Learning precisamos entender que os dados que dispomos são lidos de uma maneira muito diferente da qual visualizamos.

* **Datasets:** São conjuntos de dados estruturados utilizados como dado de entrada para treinamento de modelos de aprendizado de máquina, as linhas correspondem ao que chamamos de samples/amostras, enquanto as colunas são os atributos/features do nosso conjunto. Estes dados são representados a partir de matrizes $m$x$n$. Por isso, Álgebra Linear é importante quando aprendemos ML.

* **Features:** São nossas variáveis independentes, ou atributos, que são utilizados para realizar a previsão/classificação do dado disponibilizado. As features são as colunas que carregam informações que descrevem nosso objeto (ex: altura, peso, custo).
* **Label:** É a variável dependente, target/variável alvo, a qual desejamos descobrir o próximo valor ou classificar. É representada geralmente pela letra minúscula $y$, enquanto as features são representadas por $X$ maiúsculo.

## **Aprendizado Supervisionado**

No aprendizado supervisionado temos conhecimento a priori dos dados que vão ser utilizados, também dizemos que os rótulos são conhecidos. E é a partir deste conhecimento prévio que realizamos a decisão de qual tipo de modelo iremos aplicar sobre os nossos dados. Dentre as categorias de modelos temos dois usuais, são estes:

**Os modelos de regressão:** Utilizado quando queremos prever um valor contínuo (numérico), como preço, temperatura ou idade.

* Alguns algoritmos de regressão:
    * Simple/Multiple Linear Regression
    * Random Forest Regressor
    * Gradient Boosting Regressor

**Os modelos de classificação:** Utilizado quando queremos classificar um dado em categorias discretas. As saídas podem ser binárias (duas classes) ou multiclasse (três ou mais classes, como classificar dígitos de 0 a 9 ou tipos de flores).

* Alguns algoritmos de classificação:
    * Random Forest Classifier
    * Logistic Regression
    * Support Vector Machine (SVM)

## **Aprendizado Não Supervisionado**

Diferente do aprendizado supervisionado, no aprendizado não supervisionado a máquina não possui rótulos que indiquem a resposta correta. O algoritmo deve identificar padrões, estruturas ou relações nos dados sem orientação prévia sobre o que procurar.

O aprendizado não supervisionado subdivide-se em diferentes categorias conforme o objetivo:

**Clustering (Agrupamento):**

* Objetivo: Agrupar amostras similares em clusters com base na proximidade de características
* Algoritmos: K-means, DBSCAN, Hierarchical Clustering

**Redução de Dimensionalidade:**

* Objetivo: Comprimir a representação dos dados mantendo a maior quantidade possível de informação/variância
* Algoritmos: PCA (Principal Component Analysis), t-SNE, UMAP

**Estimação de Densidade:**

* Objetivo: Modelar a distribuição de probabilidade dos dados
* Algoritmos: Kernel Density Estimation (KDE), Gaussian Mixture Models (GMM)

**Detecção de Anomalias:**

* Objetivo: Identificar observações que desviam significativamente do padrão geral
* Algoritmos: Isolation Forest, One-Class SVM

## **Aprendizado Semi-supervisionado**

O Aprendizado Semi-supervisionado combina uma pequena quantidade de dados rotulados com um grande volume de dados não rotulados. Esta abordagem visa melhorar a capacidade de generalização do modelo em cenários onde a obtenção de dados brutos é barata, mas a rotulação manual exige especialistas e é financeiramente custosa.
O mecanismo fundamental dessa abordagem consiste em utilizar a estrutura ou distribuição dos dados não rotulados para inferir informações úteis. Os dados não rotulados auxiliam na modelagem da distribuição P(X), que por sua vez contribui para uma estimativa mais precisa de P(Y|X).

## **Aprendizado por Reforço**

O aprendizado por reforço opera de forma distinta dos paradigmas supervisionado e não supervisionado. Nesta abordagem, um agente interage com um ambiente através de um ciclo contínuo de observação, ação e recompensa.
O processo funciona da seguinte forma: o agente observa o estado atual do ambiente, seleciona uma ação com base em uma política (π) — que é a função ou regra que mapeia estados para ações — e recebe uma recompensa numérica como feedback do ambiente. O ambiente então transita para um novo estado, e o ciclo se repete.
O objetivo do agente é maximizar a recompensa acumulada ao longo do tempo, denominada retorno esperado. Para atingir este objetivo, o agente atualiza suas estimativas de valor — funções que quantificam quão benéfico é estar em determinado estado ou executar determinada ação em um estado — com base nos sinais de recompensa recebidos durante a interação com o ambiente.

## **Overfitting e Underfitting**

Overfitting e Underfitting referem-se à capacidade de generalização do modelo, analisando a discrepância de desempenho entre os dados vistos durante o treinamento e os dados não vistos durante o teste.

* **Overfitting:** Quando o modelo tem um grau de complexidade maior em relação à quantidade de dados disponíveis, fazendo com que o modelo passe a aprender tanto o padrão real quanto o ruído estocástico (aleatoriedade) presente no conjunto de treinamento, esses modelos possuem alta variância, percebemos o **overfitting** a partir do resultado percentual elevado durante o treinamento, e o baixo durante o teste com novos dados.

* **Underfitting:** Ocorre quando o modelo é excessivamente simples para capturar a estrutura subjacente dos dados. Matematicamente, dizemos que o modelo possui Alto Viés (BIAS). Ele é incapaz de modelar a relação entre as features ($X$) e os labels ($y$), ignorando padrões importantes, e fazendo com que o modelo apresente uma taxa de erro elevada tanto no conjunto de treino quanto no conjunto de teste.

#### **Tratamento de variáveis categoricas.**

* **Label Encoding:** É interessante para variáveis ordinais, pois irá atribuir um número inteiro para cada categoria, como ele categoriza de forma ordenada numa única coluna, tem chance do modelo entender que 1 > 0 e isto impactar nos resultados do modelo em questão.
* **One-Hot Encoding:** Cria uma coluna para cada categoria de dado existente no grupo selecionado, tendo aplicabilidade melhor para variáveis nominais.

#### **Escalonamento de Features**
Escalonamento é um processo que transforma os valores das features em uma escala comum, sem distorcer as diferenças nas faixas de valores nem perder informação. Isso se deve ao fato de que alguns modelos tem sensibilidade maior a escala, com isso podem ter resultados tendenciosos caso não tratado o problema, dai surge o Escalonamento de Features.

* **Padronização (StandardScaler):** Um método sensível a outliers que busca padronizar as escalas a partir da fórmula $$z = \frac{x - \mu}{\sigma}$$

* **Normalização (MinMaxScaler):** Um método muito sensível aos outliers e que busca normalizar as escalas a partir da fórmula $$x_{norm} = \frac{x - x_{min}}{x_{max} - x_{min}}$$
* **Escalonamento Robusto (RobustScaler):** Método não impactado por outliers, tem como fórmula $$x_{scaled} = \frac{x - Q_2}{IQR}$$

---

Contudo, como estamos trabalhando com um modelo de classificação em árvores, não vamos precisar realizar o escalonamento do modelo. Geralmente os modelos que precisam de escalonamento são modelos que trabalham calculando a distância entre pontos(KNN, SVM) ou utilizam gradiente para otimizações (Regressão Logística, Redes Neurais) que são afetados diretamente pela diferença escalar.

Além disto, é muito **importante** citar que o escalonamento deve ser feito apenas após a divisão dos conjuntos de teste/treino para evitar data leakage pois envolve tratamentos estatísticos, como a utilização de médias, medianas e quartis por exemplo.

#### **Divisão dos dados**

Antes de iniciar tratamentos estatísticos para preenchimento de dados nulos como por exemplo a idade, e o deck que a pessoa reside, precisamos dividir nossos dados de treino e teste para só então aplicar estes tratamentos.

Isso se deve pelo fato de que quando tiramos a média dentro do conjunto completo, fazemos com que o modelo subentenda que existem conexões entre os conjuntos de teste e treino, e assim o mesmo tenha um desempenho maior a partir disto, logo se evita realizar tratamentos antes da separação dos dados.

Pois a média de um conjunto que contêm 20% dos dados é diferente daquele conjunto que possui 80%

````python
#Divisão dos dados de teste e treino.
#É importante ressaltar que na divisão de dados removemos o Label para o conjunto teste.
X = df.drop(columns = ['Survived'])
y = df['Survived']

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size = 0.25, train_size = 0.75, random_state = 7)
````

Após a divisão dos dados seguimos com os tratamentos estatísticos, dessa forma evitamos o **data leakage.**

````python
#Variáveis para criação de função de imputação de decks por probabilidade da classe do passageiro.
conves_treino = X_treino[(X_treino['Pclass'].isin([1, 2, 3])) & (X_treino['Cabin'].notnull())][['Pclass', 'Deck']]
conves_treino.set_index('Pclass',inplace = True)

tabela_cruzada_treino = pd.crosstab(conves_treino.index, conves_treino['Deck'])

prob_treino = tabela_cruzada_treino.div(tabela_cruzada_treino.sum(axis = 1), axis = 0)
prob_treino #Visualizando a saída esperada.
````

````python
#Generalização dos Decks por classe.

def generalizar_decks(row, prob):
    if pd.notna(row['Deck']):
        return row['Deck']

    pclass = row['Pclass']
    probabilidades = prob.loc[pclass]
    probabilidades = probabilidades[probabilidades > 0]

    decks = probabilidades.index.tolist()
    probs = probabilidades.values

    return np.random.choice(decks, p = probs)

X_treino['Deck'] = X_treino.apply(lambda row : generalizar_decks(row, prob_treino), axis = 1)
X_teste['Deck'] = X_teste.apply(lambda row : generalizar_decks(row, prob_treino), axis = 1)

X_treino.drop(columns = ['Cabin'], inplace = True)
X_teste.drop(columns = ['Cabin'], inplace = True)

X_teste.info()
````

````python
#Generalização da idade com base em contagem familiar.

medias_familia_treino = X_treino.groupby('Family')['Age'].mean()
X_treino['Age'] = X_treino['Age'].fillna(X_treino['Family'].map(medias_familia_treino))
X_teste['Age'] = X_teste['Age'].fillna(X_teste['Family'].map(medias_familia_treino))

#Apesar de não termos blocos de código, foi identificado que todas as variáveis com família igual a 10, possuem idade nula.
#Logo para tratar isso será feito a generalização a partir da média de todo o conjunto de dados.

media_treino = X_treino['Age'].mean()

X_teste.loc[X_teste['Family'] == 10, 'Age'] = media_treino
X_treino.loc[X_treino['Family'] == 10, 'Age'] = media_treino
X_treino.info()
````

````python
#Neste bloco realizo a normalização dos dados da coluna 'Deck' com uma função do Pandas.
#Irá haver perda de alguns dados de acordo com a divisão dos modelos, pois caso não haja casos de um dos Decks em treino e haja em teste, o dado será removido do Teste.
#Contudo, como haverão poucos casos, seguirei com este tipo de tratamento apenas para ilustrar o que deve ser feito.

X_treino = pd.get_dummies(X_treino, columns = ['Deck'], drop_first = True)
X_teste = pd.get_dummies(X_teste, columns = ['Deck'], drop_first = True)

X_teste = X_teste.reindex(columns = X_treino.columns, fill_value = 0)
````

### **Execução do modelo e validação, uma via de mão dupla.**

Após dividir nosso conjunto de dados em Treino/Teste, passamos para a fase de implementação do modelo, onde iremos colocar estes dados tratados à prova, para só então realizarmos as validações e verificarmos sua eficácia.

````python
#Execução dos modelos selecionados.
cross_validation = StratifiedKFold(n_splits = 6, shuffle = True, random_state = 7)

metricas = ['f1', 'precision', 'recall', 'accuracy']

modelos = {
    'Regressão Logistica' : LogisticRegression(max_iter = 1000, random_state = 7),
    'Random Forest' : RandomForestClassifier(max_depth = 5, random_state = 7),
    'Gradient Boosting' : GradientBoostingClassifier(max_depth = 5, random_state = 7)
}

for nome, modelo in modelos.items():
  scores = cross_validate(modelo, X_treino, y_treino, cv = cross_validation, scoring = metricas)
  print('-' * 30)
  print(f"Modelo: {nome}\nF1: {scores['test_f1'].mean():.4f} (+/- {scores['test_f1'].std():.4f})")
  print(f"Precisão: {scores['test_precision'].mean():.4f} (+/- {scores['test_precision'].std():.4f})")
  print(f"Recall: {scores['test_recall'].mean():.4f} (+/- {scores['test_recall'].std():.4f})")
  print(f"Acurácia: {scores['test_accuracy'].mean():.4f} (+/- {scores['test_accuracy'].std():.4f})")
````

Acima utilizamos de validação cruzada dividindo o nosso dataset de treino em 'n' divisões a partir do método StratifiedKFold e então rodando todos os modelos e comparando-os a partir de algumas métricas que venhamos a desejar visualizar para definir qual será o melhor modelo.

E após finalizarmos esta validação cruzada, finalmente podemos seguir para o treinamento e teste do modelo, tendo em mente os resultados que obtivemos anteriormente.

**DICA:** Temos como testar vários ajustes de hiperparâmetros no Scikit-Learn, logo podemos ver qual terá um melhor resultado, contudo é muito importante levar em consideração que podemos causar tanto overfitting quanto underfitting, logo há necessidade de ter **cuidado.**

````python
modelo = RandomForestClassifier(max_depth = 5, random_state = 7) #Invocação do modelo com as configurações setadas conforme processo anterior.
modelo.fit(X_treino, y_treino) #Treinamento do modelo.
y_pred = modelo.predict(X_teste) #Teste com a partição não conhecida.
y_pred_treino = modelo.predict(X_treino)
````

````python
print(classification_report(y_teste, y_pred, target_names=['0', '1'])) #Visualização geral dos acertos das respostas em teste com o predito.
````

````python
print(f'Acurácia: {accuracy_score(y_teste, y_pred)}')
print(f'F1_Score (Treino): {f1_score(y_treino, y_pred_treino)}') #Verificar possibilidade de overfitting comparando os F1s de treino/teste. Caso treino muito maior que teste, há overfitting.
print(f'F1_Score (Teste): {f1_score(y_teste, y_pred)}\n\n')

cm = confusion_matrix(y_teste, y_pred)
print(f'Matriz da confusão\n{cm}')
````

* **(FP - Falso Positivo):** O modelo preveu que 34 pessoas sobreviveram, quando na realidade não sobreviveram.

### **Modelo com Overfitting? O que fazer?**

No caso deste material de estudo, temos o modelo com Overfitting e isso se deve pelo fato de que a diferença percentual entre o F1-Score de treino/teste é de aproximadamente 15%, representando sim Overfitting durante o treino.

Para tratar isto, podemos buscar ajustes nos hiperparâmetros visando identificar qual trará o melhor resultado e menor diferença percentual entre pontuações, e a partir disto aplicarmos as modificações.

Além disto também é normal, durante a troca de hiperparâmetros outros modelos se tornarem melhores que outros previamente escolhidos, com isto encerramos nosso estudo com o documento de métricas de avaliação.

## **Métricas de avaliação**

#### **Função de Custo vs. Métrica de Avaliação**

Esta distinção é fundamental:

* **Função de Custo (Loss Function):** Função matemática que o algoritmo de otimização minimiza durante o treinamento. Deve ser diferenciável para permitir gradiente descendente. Exemplos: MSE, Cross-Entropy Loss.

* **Métrica de Avaliação:** Medida utilizada após o treinamento para interpretar o desempenho do modelo em termos compreensíveis ao negócio. Não precisa ser diferenciável. Exemplos: Acurácia, F1-Score, R².

Em alguns casos, a mesma função pode servir ambos propósitos (MSE é loss e métrica), mas nem sempre (Acurácia não é diferenciável, portanto não serve como loss).

## **Métricas para Aprendizado Supervisionado**

A avaliação consiste em quantificar a concordância entre o valor previsto e o valor real.

### **Classificação**

#### **Métricas Essenciais (Prioridade de Aprendizado)**

Estas métricas derivam da **Matriz de Confusão**, que categoriza as previsões em:

* **TP (True Positive):** Previu positivo, era positivo

* **TN (True Negative):** Previu negativo, era negativo

* **FP (False Positive):** Previu positivo, era negativo (Erro Tipo I)

* **FN (False Negative):** Previu negativo, era positivo (Erro Tipo II)

**Acurácia:**

* Fórmula: (TP + TN) / (TP + TN + FP + FN)

* Quando utilizar: Classes balanceadas, custo de erros simétrico

**Precisão:**

* Fórmula: TP / (TP + FP)

* Quando utilizar: Custo alto de Falso Positivo (ex: classificação de spam)

**Recall (Sensibilidade):**

* Fórmula: TP / (TP + FN)

* Quando utilizar: Custo alto de Falso Negativo (ex: diagnóstico de câncer)

**F1-Score:**

* Fórmula: 2 × (Precision × Recall) / (Precision + Recall)

* Quando utilizar: Necessidade de balanço entre Precisão e Recall

**AUC-ROC:**

* Definição: Área sob a curva ROC

* Quando utilizar: Comparação de modelos independente de threshold

#### Métricas Avançadas (Casos Específicos)

**Baseadas na Matriz de Confusão:**

* **Especificidade:** TN / (TN + FP) — Capacidade de identificar negativos reais

* **F-Beta Score:** Generalização do F1 com peso β para priorizar Recall (β > 1) ou Precisão (β < 1)

* **MCC (Matthews Correlation Coefficient):** Varia de -1 a +1, mais robusto para classes desbalanceadas

* **Cohen's Kappa:** Concordância descontando acerto aleatório

* **Balanced Accuracy:** Média entre Sensibilidade e Especificidade

**Baseadas em Probabilidade (Threshold-Independent):**

* **AUC-PR:** Área sob curva Precision-Recall, preferível para classes muito desbalanceadas

* **Gini Coefficient:** 2 × AUC - 1

* **Log-Loss:** Penaliza divergência entre probabilidade predita e classe real (também usada como loss function)

* **Brier Score:** Erro quadrático médio das probabilidades

**Para Classificação Multilabel:**

* **Hamming Loss:** Fração de rótulos incorretos

* **Jaccard Score:** Interseção sobre união dos rótulos

* **Exact Match Ratio:** Requer acerto de todos os rótulos

---

### **Regressão**

#### **Métricas Essenciais**

**MAE (Mean Absolute Error):**

* Fórmula: (1/n) × Σ|yi - ŷi|

* Interpretação: Erro médio na escala original

**MSE (Mean Squared Error):**

* Fórmula: (1/n) × Σ(yi - ŷi)²

* Interpretação: Penaliza erros grandes mais severamente

**RMSE (Root Mean Squared Error):**

* Fórmula: √MSE

* Interpretação: MSE na escala original

**R² (Coeficiente de Determinação):**

* Fórmula: 1 - (SSres / SStot)

* Interpretação: Proporção da variância explicada (0 a 1)

#### **Métricas Avançadas**

**Erros Relativos:**

* **MAPE:** Erro médio percentual — problemático quando y se aproxima de zero

* **SMAPE:** MAPE simétrico, mais estável próximo a zero

* **MSLE:** Penaliza subestimativas mais que superestimativas

**Variância:**

* **Adjusted R²:** R² corrigido pelo número de features, penaliza complexidade desnecessária

**Quantis:**

* **Pinball Loss:** Para regressão quantílica (previsão de intervalos)

---

### **Métricas para Aprendizado Não Supervisionado**

A validação é mais complexa pela ausência de rótulos verdadeiros.

### **Clustering**

#### **Validação Interna (Sem Rótulos - Baseada em Geometria)**

Utilizada quando não existe ground truth para comparação:

**Silhouette Score:**

* Intervalo: -1 a 1

* Objetivo: Maximizar (coesão interna, separação externa)

**Davies-Bouldin Index:**

* Intervalo: 0 a ∞

* Objetivo: Minimizar

**Calinski-Harabasz Index:**

* Intervalo: 0 a ∞

* Objetivo: Maximizar

#### **Validação Externa (Com Rótulos de Referência)**

Utilizada quando existe um gabarito oculto para comparação:

* **Adjusted Rand Index (ARI):** Similaridade entre clusterizações, ajustado para acaso (-1 a 1)

* **Normalized Mutual Information (NMI):** Informação compartilhada normalizada (0 a 1)

* **V-Measure:** Média harmônica entre Homogeneidade e Completude

#### **Redução de Dimensionalidade**

* **Reconstruction Error:** Erro entre entrada original e reconstrução (PCA, Autoencoders)

* **Explained Variance Ratio:** Proporção da variância retida por componente

* **Trustworthiness:** Preservação da vizinhança local após projeção (t-SNE, UMAP)

---

### **Métricas para Aprendizado Semissupervisionado**

A avaliação final utiliza métricas supervisionadas sobre um conjunto de teste rotulado. Adicionalmente, monitoram-se métricas de processo:

* **Pseudo-Label Accuracy:** Acurácia dos rótulos atribuídos automaticamente

* **Confidence Distribution:** Distribuição das probabilidades preditas (modelos confiantes concentram massa próxima a 0 e 1)

* **Retention Rate:** Proporção de dados não rotulados que superaram o limiar de confiança

---

### **Resumo: Escolha da Métrica por Contexto**

**Classes balanceadas:**

* Métricas recomendadas: Acurácia, F1

**Classes desbalanceadas:**

* Métricas recomendadas: AUC-PR, MCC, Balanced Accuracy

**Custo assimétrico de erros:**

* Métricas recomendadas: Precisão ou Recall (conforme o caso)

**Comparação de modelos:**

* Métricas recomendadas: AUC-ROC, Cross-Validation Score

**Regressão com outliers:**

* Métricas recomendadas: MAE, Huber Loss

**Regressão padrão:**

* Métricas recomendadas: RMSE, R²

**Clustering exploratório:**

* Métricas recomendadas: Silhouette, Elbow Method
