<center><h3><b>Do supervisionado ao não supervisionado</b></h3></center>

Como abordado no tópico anterior, o aprendizado de máquina é uma ferramenta muito valiosa e tem vários meios de utilização a depender dos dados que são disponibilizados durante o processo de execução do algoritmo escolhido, e é antes de definir o algoritmo que utilizaremos que precisamos identificar qual é o método de aprendizado que mais se adequa a situação-problema apresentada para solução:
* <b>Aprendizado supervisionado:</b>
    * Este tipo de aprendizado tem um enfoque maior em auxiliar na aplicação de regressões e também classificações de comportamentos referentes a dados diversos. Funcionando a partir de dados rotulados, esses dados vão conter características de algo que são utilizadas para formar conjuntos de treinamento e teste para o algoritmo, sendo o de teste: Sem resposta, e o de treino: Com algumas respostas para "n" características. A ideia por trás deste algoritmo está em identificar estes padrões e transformar estes numa resposta de qual elemento tem essas características, ou com base nesses valores o quanto minhas ações podem subir por exemplo.
* <b>Aprendizado não-supervisionado:</b>
    * Já esta metodologia de aprendizado não tem dados rotulados, então por não depender da disponibilização de informações para trabalhar, a ideia deste aprendizado é identificar padrões semelhantes e não semelhantes visando classificar estes em grupos que chamamos de clusters. Um algoritmo bem conhecido é o K-means, que faz este trabalho de segregar grupos distintos e unir grupos de mesmas características. Sua usabilidade está presente em várias áreas e inclusive auxilia na identificação de tumores e outros fatores.
* <b>Aprendizado por reforço</b>
    * Este método por sua vez tem como base a aprendizagem por metas. O aprendizado é baseado na experiência adquirida por aquele modelo após "n" períodos, sendo mais utilizado para casos em que há imprevisibilidade e variância elevada dos dados, permitindo que a máquina lide com erros e acertos através do seu processo de aprendizado, permitindo que a abordagem correta seja aplicada para entrega de resultados precisos.

<h2><b><center>Testes paramétricos e não paramétricos</center></b></h2>

É importante entender que é de extrema importância o conhecimento estatístico para construir algorítmos de aprendizado de máquina. E com isto, temos os testes paramétricos e não paramétricos que são responsáveis por nos auxiliar na criação de inferências sobre uma população a partir de uma amostra desta, contudo de acordo com a distribuição dos dados existem diferentes tipos de testes paramétricos e não paramétricos que podem ser utilizados.

* <b>Teste paramétrico:</b> Este tipo de teste precisa que algumas condições sejam cumpridas para que possa ser utilizado, dentre elas:
    * A variável dependente deve ser distribuida de forma aproximadamente normal na população.
    * A variância entre os grupos analisados não pode ser muito distinta (heteroscedasticidade), para identificar este ponto temos o teste de Levene.
    * Os resultados do indivíduo ou unidade da amostra devem ser independentes entre si.
    * Testes paramétricos são mais indicados para variáveis contínuas e discretas, caso nominal ou ordinal é indicado o teste não paramétrico.

        * **Exemplos de testes paramétricos:**
            * Teste t de Student: O teste t é usado para comparar as médias de dois grupos. Existem diferentes tipos, dependendo da natureza dos dados
            * Análise de Variância (ANOVA): A ANOVA é usada para comparar as médias de três ou mais grupos.
            * Coeficiente de Correlação de Pearson: Este teste mede a força e a direção da relação linear entre duas variáveis contínuas
            * Regressão linear: A regressão linear é usada para modelar a relação entre uma variável dependente e uma ou mais variáveis independentes, com o objetivo de prever resultados.
            * Teste Z: O teste Z é usado para comparar a média de uma amostra com uma média populacional conhecida, mas é mais apropriado quando se conhece o desvio padrão da população e a amostra é grande. 

     
* <b>Teste não paramétrico:</b> Os testes não paramétricos existem para analisar dados que não atendem às suposições estritas dos testes paramétricos, principalmente a suposição de que os dados seguem uma distribuição normal. Os testes não paramétricos trbalham com a ordenação dos dados (postos ou ranks), ou seja, a posição de cada observação em relação às outras, graças a isto estes métodos de análise se tornam menos dependentes da distribuição dos dados, já que a análise se baseia na ordem relativa e não nos valores absolutos dos dados, por exemplo. Logo geralmente usamos os testes não paramétricos quando:
    * Quando a distribuição se desvia significamente da forma de sino.
    * Com amostras pequenas que possuem menos de 30 unidades, pois é dificil determinar se de fato a distribuição é normal pelo baixo número de unidades presentes na amostra.
    * Quando há dados ordinais ou nominais para análise.
    * Quando a robustez é importante, já que os testes não paramétricos sõa menos sensíveis a valores discrepantes (outliers).

        * **Exemplos de testes não paramétricos:**
            * Teste de Mann-Whitney U: Equivalente ao teste t de Student para amostras independentes.
            * Teste de Wilcoxon: Equivalente ao teste t pareado (para amostras dependentes).
            * Teste de Kruskal-Wallis: Equivalente à Análise de Variância (ANOVA).
            * Correlação de Spearman: Equivalente à correlação de Pearson.

<h2><b><center>Modelos paramétricos e não paramétricos</center></b></h2>

Agora considerando o aprendizado de máquina, os modelos paramétricos assumem que há uma função conhecida, com isto é pressuposto que entendamos como os nossos dados se comportam, e como neste modelo é esperado o conhecimento inicial sobre a distribuição dos dados, o mesmo se torna mais simples de entender. Alguns exemplos destes modelos paramétricos são:

* Regressão linear.
* Regressão logística.
* Redes neurais simples.

Enquanto isso, os modelos não paramétricos são utilizados quando não há fortes suposições sobre como os nossos dados se comportam ou funções de mapeamento da qual podemos utilizar. Logo é o mesmo que querer utilizar da regressão linear para predizer o valor de venda de diversas casas em regiões diferentes do Brasil, sendo que existem vários micro e macro fatores que impactam diretamente no valor de venda. Com isto, não seria a melhor decisão a escolha de uma função linear. Com isto poderiamos utilizar do algoritmo de árvores de decisão que se enquadraria melhor aos nossos dados. Portanto para modelos não paramétricos temos alguns exemplos:

* Árvores de decisão
* Support Vector Machine (SVM)

Aprofundando primeiramente em um modelo de aprendizado de máquina simples que é supervisionado, temos a regressão linear da qual tem como objetivo prever ou predizer valores de determinada variável dependente a partir da visão de uma variável independente. A regressão linear pode ter mais de uma variável independente? Sim! Essa seria uma regressão linear múltipla, contudo caso haja apenas uma, seria uma regressão linear simples.

<h2><b><center>Um pouco sobre as redes neurais</center></b></h2>

Seguindo agora para redes neurais, a menção é importante pelo fato de necessitarmos desse conhecimento para algoritmos de aprendizado profundo. E basicamente, as redes neurais artificiais surgiram como um método de imitar a rede neural humana de forma computacional, tudo advindo da pergunta "Será que as máquinas conseguem ter autonomia e pensar por si próprias?", e portanto foram criadas diversos tipos de redes para tentar se aproximar o máximo possível da humana.

<i>Contudo, antes de avançarmos para apresentação de cada uma das redes neurais, vamos explicar como funcionam estas redes neurais.</i>

As redes neurais são modelos de computação que simulam o funcionamento da rede neural cerebral humana de forma que cada neurônio/camada/dimensão dessa rede se comuniquem entre si para gerar uma espécie de saída/resultado. Dentro da rede neural computacional existem diversas camadas que podem ser de entrada, camadas ocultas ou de saída, sendo que as camadas ocultas consistem em unidades que transformam a entrada em algo que a camada de saída possa utilizar de alguma forma.

Por fim, abaixo teremos explicações sobre algumas redes neurais:

---

<h4><b><center>Rede neural Feed-Forward</center></b></h4>

Essa arquitetura é bem simples, basicamente a informação viaja da entrada para a saída sem arrodeios. O máximo que está pode ter durante o caminho da entrada para a saída são as camadas ocultas, contudo se houver mais de uma camada oculta essa rede passaa ser chamada de rede neural profunda.

<h4><b><center>Rede neural recorrente (RNNs)</center></b></h4>

As redes neurais recorrentes são redes que dividem seu trabalho em diversas camadas ocultas das quais tentam identificar qual é a próxima palavra dentro de uma frase incompleta, a partir disto ela tenta dezenas ou até mesmo milhares de palavras na tentativa de encaixar o que combina e faz mais sentido com aquele contexto. Simulando de fato a rede neural humana, pois quando ouvimos um "Era uma vez" supomos geralmente que a próxima palavra é "um/uma", basicamente lea faz o mesmo. Sendo uma rede neural muito usada no NLP e no reconhecimento de fala, por exemplo.

<h2><b><center>Um pouco mais sobre o classificador Perceptron</center></b></h2>

O perceptron foi uma ideia introduzida por Frank Roseblatt em 1957, ele propôs uma regra de aprendizado do Perceptron baseado no neurônio MCP criado por McCulloch e Walter Pitts, e basicamente esse neurônio MCP imitaa funcionalidade biológica de um neurônio, sendo também considerado um neurônio artifícial. Portanto, sendo este um neurônio artificial, aceita-se entradas binárias e devolve em saídas binárias (neste caso em específico do Perceptron e MCP, contudo não se restringe a eles), então o Perceptron se tornou um algoritmo usado principalmente para problemas de classificação binária e pode ser considerado um algoritmo de aprendizagem supervisionada de classificação binária.

É importante mencionar que existem dois tipos de perceptrons: **os de camada única e os de camada múltipla:**
* **Camada única:** Perceptrons de camada única podem aprender apenas padrões linearmente separáveis.
* **Camada múltipla:** Perceptrons multicamadas ou redes neurais feedforward com duas ou mais camadas têm maior poder de processamento.
 
<h4><b><center>Regra de aprendizado do Perceptron, suas entradas e características principais</center></b></h4>

Agora falando um pouco das regras de aprendizado do perceptron, é afirmado que o algoritmo aprende automaticamente os coeficientes de peso ideais. Com isto, os recursos de entrada são multiplicados com os pesos identificados para determinar se o neurônio irá disparar ou não. E bom, o perceptron ele recebe vários sinais de entrada, e é quando a soma destes sinais de entrada excedem um certo limite, que ele emite um sinal ou não retorna uma saída. No contexto de aprendizagem supervisionada de classificação, isso pode ser usado pra prever a classe de uma amostra.

**Sobre a entrada:** O perceptron vai aceitar entradas, modera-las com certos valores de peso, e em seguida aplicar a função de transformação que irá gerar o resultado final.

**Por fim suas características principais:**

* É um algorítmo de aprendizagem supervisionada de classificadores lineares binários de camada única ou múltipla.
* Os coeficientes de peso ideais são aprendidos automaticamente.
* Os pesos são multiplicados com os recursos de entrada e a decisão é feita se o neurônio é disparado ou não.
* A função de ativação aplica uma regra para verificar se a saída da função de ponderação é maior que zero. É importante ressaltar este ponto poisé a função de ativação que informa se o neurônio será ou não ativo, determinando se aquela informação é relevante ou não, e quando for, ignorar o resultado gerado.
* O limite de decisão linear é traçado permitindo a distinção entre as duas classes separáveis linearmente +1 e -1.
* Se a soma dos sinais de entrada exceder um certo limite, o Perceptron emitirá um sinal. Caso contrário, não há saída.
* Os tipos de funções de ativação incluem as funções de sinal, degrau e sigmóide.

O algoritmo de Backpropagation é uma forma de cálculo derivadas veloz, tendo dois passos principais:
* **Forward pass:** Ele permite que as entradas passem por toda a rede considerando as previsões e resultados.
* **Backward pass:** Ele realiza o gradiente para realizar a aplicação de forma recursiva para realizar a atualização dos pesos da rede especificada.
