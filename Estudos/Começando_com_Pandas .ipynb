### **Filtragem**

A partir desta seção temos um pouco mais de complexidade envolvida, pois conseguimos realizar todos os três metodos juntos para trazer resultados satisfatórios, então a parte de filtragem é sobre coletar um dado a partir de 'n' condições estabelecidas como por exemplo:

````python
#Condições simples.
bahia = x[x['uf_acidente'] == 'BA']
````

Quando realizamos uma condição de filtragem em Pandas, somos obrigados a chamar o dataframe e então abrir colchetes para chamá-lo novamente e só então realizarmos a filtragem. Isso se dá pelos seguintes fatos:

* **Chamado interno:** É onde formulamos nosso filtro e criamos uma mascara booleana, nela teremos os nosso valores categorizados como **True/False.**
* **Chamado externo:** O chamado externo é o primeiro chamado do DataFrame e a parte em que não há filtros, nela fazemos o algoritmo entender que queremos trazer os dados do DataFrame que condizem com a mascara criada internamente.

````python
#Destrinchando o mesmo filtro temos:

bahia = x[ #Aqui dentro armazenamos a mascara, nosso filtro. E fora disto temos o chamado externo para que os dados não venham como Booleano.
          x['uf_acidente'] == 'BA']


bahia = x['uf_acidente'] == 'BA' #Mascara Booleana
bahia.unique().tolist() #Além disto, quando filtramos desta forma trazemos apenas a Series/Coluna do filtro, para trazer mais além desta columa
                        #Se faz necessário a aplicação da mascara criada no dataframe original, ou um filtro pré-estabelecido de quais são as colunas desejadas.
````

````python
#Agora quando falamos de condições complexas, temos uma grande gama de filtros que podem ser adicionadas, como por exemplo:

amazonas = x[(x['uf_acidente'] == 'AM') &
             (x['mes_ano_acidente'].isin([12018, 82018])) &
             (x['genero'] == 'DESCONHECIDO') &
             (x['faixa_idade'] == 'ENTRE 50 E 54 ANOS')] #No Pandas é obrigatório a utilização dos parenteses para filtros múltiplos.
amazonas
````

Por fim, abaixo deixo alguns dos outros condicionais ou meios de filtro:
* **( | ):** Um simbolo de pipe referenciam a condicional **OR.**
* **( & ):** Um simbolo de **&** equivale a condicional **AND.**
* **( ~ ):** Um simbolo de **~** equivale a condicional **NOT.**
* ```.isin([])```: Recebe listas de valores para verificar se um dado 'x' está dentro dos valores presentes na lista, esse método é utilizado ao lado da seleção da coluna em questão que passará pela filtragem
* ```Sinais de comparação```: Como por exemplo ```>```, ```>=```, ```<```, ```<=```, ```==```, ```!=```.

## **Manipulação e Transformação de Colunas**

As vezes alguns dados que precisamos não existem e precisam ser criados, a partir disto conseguimos criar por exemplo colunas calculadas de um valor como por exemplo o CPM, CPC que são métricas de Performance no Marketing.

````python
df.drop(columns = ['codigo_sinistro'], inplace = True)
df.columns.tolist()
````

Além disto podemos usar tanto a função .map() quanto a função .apply() para realizar operações nas coluna tanto com funções quanto com condições.

## **Agregação e Agrupamento (GroupBy)**

Caso venhamos a trabalhar com campos calculados e queiramos por algum motivo resumir um resultado, podemos realizar o agrupamento e agregação dessas dados pelo Pandas, a agregação é lida como ```.agg()``` e consegue aplicar diversos métodos incluindo média (mean), somatória de valores (sum), dentre outros. Além disso para agrupar usamos ```.groupby()```.

````python
raw_data = pd.read_csv(r'/content/sample_data/california_housing_train.csv')
x = raw_data['housing_median_age'].unique().tolist()
print(min(x), max(x))

raw_data['housing_median_age'] = raw_data['housing_median_age'].apply(lambda x: '1 a 20' if x <= 20 else ('21 a 35' if x > 20 and x <= 35 else '36 and 50'))
raw_data.info()
````

````python
raw_data = raw_data.groupby('housing_median_age', as_index = False)['median_house_value'].agg('mean')
raw_data
````

## **Combinação de Múltiplos DataFrames**

As vezes trabalhamos com múltiplos dataframes com dados similares, ou que são divididos por algum tipo de produto e acabam que por possuir uma chave em comum. Nisto, para analisarmos este dado com maior exatidão, ou até mesmo realizar agregações com outros valores de investimento e venda, realizamos a junção destes dados.


````python
#As vezes são dois dataframes iguais, contudo com linhas diferentes. Neste caso:
x = pd.read_csv(r'/content/sample_data/california_housing_train.csv')
y = pd.read_csv(r'/content/sample_data/california_housing_test.csv')
print('Linhas do primeiro DataFrame:', x.shape[0], '\nLinhas do segundo DataFrame:', y.shape[0])

z = pd.concat([x, y])
print('Após concatenação dos valores dos DataFrames: ', z.shape[0])
````

````python
#Enquanto as vezes temos diversos datasets que se complementam e possuem uma chave que pode ser utilizada em sua conexão.
x = pd.read_csv(r'/content/sample_data/Vitimas_DadosAbertos_20251112.csv', sep = ';')
y = pd.read_csv(r'/content/sample_data/TipoVeiculo_DadosAbertos_20251112.csv', sep = ';')

veiculos = y.columns.to_list()
vitimas = x.columns.to_list()
for colunas in veiculos:
  if colunas in vitimas:
    print('Coluna com mesmo nome encontrada:', colunas)
  else: pass
````

````python
#Após isso podemos seguir diretamente pro Merge que verificará se de fato contemos dados iguais e trará os dados com base nisto.

correspondencias = pd.merge(left = x, right = y, how = 'outer', indicator = True) #Aqui eu consigo verificar quantas linhas deram Match de informações e quantas não.
right_only = correspondencias[correspondencias['_merge'] == 'right_only']
left_only = correspondencias[correspondencias['_merge'] == 'left_only']
both = correspondencias[correspondencias['_merge'] == 'both']

print(f'Dados existentes apenas em Veículos: {right_only["_merge"].count()}\n Dados existentes apenas em Vítimas {left_only["_merge"].count()}\n Dados existentes em ambos {both["_merge"].count()}')
````

## **Introdução a Dados Temporais (Datetime)**

Quando falamos de dados temporais, falamos de um dado de tempo, uma data completa, um timestamp, ou um objeto categórico em formato de data, agora como trabalhar com isso? O que fazer pra extrair partes específicas?

````python
#Conversão para datetime
#Além do astype('datetime64[ns]') temos:

x.info()
````

````python
x['data_acidente'] = pd.to_datetime(x['data_acidente'])
x.info()
````

````python
#E agora que transformamos a data do acidente em Datetime, podemos acessar partes específicas da mesma, como por exemplo:

x['mes'] = x['data_acidente'].dt.month
x['ano'] = x['data_acidente'].dt.year
x['dia'] = x['data_acidente'].dt.day
x[['ano', 'mes', 'dia']].sample(3)
````

````python
#Assim como podemos realizar cálculos entre os períodos para descobrir por exemplo quantos dias se passaram desde então:

x['período_dados'] = x['data_acidente'].max() - x['data_acidente'].min()
x['período_dados'].unique()
````

## **Visualização Rápida de Dados**

É interessante frisar que o Pandas como dito lá em cima, tem muitas integrações dentro de diversas bibliotecas, e é exatamente por isso que ele consegue nos fornecer visualizações rápidas dos nossos dados de forma gráfica. Interessante não é mesmo?


````python
#Por exemplo, podemos seguir com um gráfico de barras, ou linhas.

x['total_acidentes'] = x.groupby('uf_acidente')['uf_acidente'].transform('count') #Diferente do agg que irá agregar, o transform nos permitirá
                                                                                  # repetir o resultado total de nosso agrupamento para os grupos iguais.

x.groupby('ano_acidente')['total_acidentes'].sum().plot(kind = 'bar')
````
