### Set Type


Um objeto Set, √© uma cole√ß√£o de dados desordenados que cont√™m diferentes objetos hashable/hashe√°veis. Ele tem como fun√ß√£o principal a deduplica√ß√£o de valores iguais, e funciona bem com opera√ß√µes matem√°ticas como intersec√ß√µes, uni√µes, diferen√ßas sim√©tricas. Al√©m de poder receber diversos tipos de valores, dentro deste t√≥pico temos dois tipos de set que funcionam de forma similar.

Set: Este √© um objeto mut√°vel e que pode sofrer altera√ß√µes, logo n√£o √© hashable, contudo, mant√™m as fun√ß√µes principais de deduplica√ß√µes. Para usar este tipo de dado basta utilizar chaves, e vale ressaltar que esse tipo de dado n√£o dados mut√°veis.

Frozenset: Este √© um objeto imut√°vel e que n√£o pode sofrer altera√ß√µes por m√©todos ap√≥s criado, com isto este tipo de dado se torna hashable permitindo que outras funcionalidades e trabalhos sejam explorados se comparado ao Set que permite adi√ß√£o de novos dados. Contudo, para criar um dado deste tipo √© necess√°rio usar o construtor como no exemplo a seguir: x = frozenset([1, 2, 3])



````python
#Podemos definir sets de duas formas. usando o set() ou usando {}.

set1 = {1, 2, 3}
set2 = {2, 3, 4}

# Opera√ß√µes matem√°ticas que podem ser utilizadas com Set:

print(f"Opera√ß√£o de Uni√£o: {set1.union(set2)}") #-> Equivalente a usar [ | ]
print(f"Opera√ß√£o de Interse√ß√£o: {set1.intersection(set2)}") #-> Equivalente a usar [ & ]
print(f"Opera√ß√£o de Diferen√ßa: {set1.difference(set2)}") #-> Equivalente a usar [ - ]
print(f"Opera√ß√£o de Diferen√ßa sim√©trica: {set1.symmetric_difference(set2)}") #-> Equivalente a usar [ ^ ]

# Frozenset como elemento de set
conj = {frozenset([1, 2]), frozenset([3, 4])}

#No Set tamb√©m podemos usar as seguintes opera√ß√µes.

set3 = set1.union(set2)
set3.add(5)
set3.remove(1)
print(set3)
set3.clear()
print(f'Valor printado: {set3}')
````

## **Imutabilidade e mutabilidade em Python**

Quando falamos de mutabilidade e imutabilidade em Python, comentamos sobre se a vari√°vel √© capaz de mudar ap√≥s ser declarada ou n√£o, contudo √© muito importante lembrar que toda vari√°vel que criamos apenas referencia o objeto como parte de sua mem√≥ria tempor√°ria, e n√£o armazena de fato aquele objeto em mem√≥ria como algumas linguagens.

Entre as vari√°veis imut√°veis temos:

* Tuple
* Frozenset
* Bytes
* Integer
* Float
* String
* Range
* Complex
* None
* Bool

Enquanto para mut√°veis temos:

* List
* Dictionary
* Bytearray
* Set

---

Quando falamos de valores imut√°veis temos o que chamamos de 'Hash' que basicamente s√£o valores dos quais nos permitem rastrear valores a partir da atribui√ß√£o de um n√∫mero inteiro que pode ou n√£o ser negativo, a ideia por tr√°s do Hash √© que ele n√£o ir√° ser alterado nunca para uma vari√°vel, a menos que um novo objeto seja criado utilizando desta.

````python
x = 'Alice in the Wonderland'
print(hash(x))
x = 'Cheshire'
print(hash(x))
#Como podemos ver se tivessemos uma vari√°vel que referencia a um nome teriamos um Hash X, enquanto uma que referencia a outro nome teriamos um hash Y
#E o que define uma vari√°vel imut√°vel?
````

Uma vari√°vel imut√°vel √© toda aquela vari√°vel que n√£o pode ser modificada ap√≥s sua cria√ß√£o, pois qualquer mudan√ßa que venha ocorrer criar√° um novo objeto por cima daquela vari√°vel, consequentemente gerando novos IDs e Hashs.

Esse caso ocorre quando tentamos ir de uma atribui√ß√£o ```x = 1``` para ```x =  2```

````python
x = 1
print('hash:', hash(x))
print('id:', id(x))
x += 2
print('hash:', hash(x))
print('id:', id(x))
````

## Programa√ß√£o orientada a objeto

O paradigma da orienta√ß√£o a objeto em Python nada mais √© que uma forma de solucionar problemas apresentados no mundo da programa√ß√£o com orienta√ß√£o a objeto, e o que seria a orienta√ß√£o a objeto nesse caso?

### Para que serve e quando utilizar orienta√ß√£o a objeto?

A orienta√ß√£o a objetos serve principlamente quando precisamos representar conceitos abstratos tanto do mundo real quanto computacional, e n√£o deve ser utilizado quando podemos resolver os problemas com fun√ß√µes simples e algumas poucas linhas de c√≥digo. √â interessante pensar que a orienta√ß√£o objeto pode ser implementada sempre que for necess√°rio pontos como:

* Encapsular estado junto ao comportamento relacionado
* Reutilizar c√≥digo atrav√©s de heran√ßa ou composi√ß√£o
* Representar m√∫ltiplas entidades com caracter√≠sticas e a√ß√µes pr√≥prias
* Manter e evoluir sistemas complexos de forma organizada
* Implementar polimorfismo para diferentes comportamentos com mesma interface

A come√ßar podemos seguir com um exemplo onde usamos fun√ß√µes para definir quantias a depositar/sacar de um usu√°rio de um banco.

````python
#Numa fun√ß√£o teriamos:

saldo = 0 #Isso seria um atributo n√£o instanciado.

def depositar(nome, valor):
  global saldo
  saldo += valor

def sacar (nome, valor):
  global saldo
  saldo -= valor

depositar('rodrigo', 500)
depositar('Anabele', 500)
print(saldo)
````

Contudo, como podemos ver o acesso da vari√°vel saldo √© f√°cilmente modificado o que nos permite visualizar o impacto que isso poderia causar para cada um destes usu√°rios que vamos chamar de inst√¢ncias, al√©m de que para o registro destes seria necess√°rio a utiliza√ß√£o de listas ou dicion√°rios aninhados para armazenamento destes valores, o que de qualquer forma poderia ainda assim ocasionar problemas e se tornar invi√°vel de acordo com a quantia de inst√¢ncias.

Agora vejamos a mesma aplica√ß√£o com a utiliza√ß√£o de classe:

````python
class Conta:
  def __init__(self, nome, idade, saldo = 0):
    self.nome = nome
    self.idade = idade
    self.saldo = saldo

  def sacar(self, valor):
    self.saldo -= valor

  def depositar(self, valor):
    self.saldo += valor

jorge = Conta('Jorge', 18)
maria_isabela = Conta('Maria Isabela', 28, 3000)
claudinho = Conta('Claudinho', 33, 3000000000000)

jorge.depositar(500) #Depositando valor na conta do Jorge :)
jorge.sacar(650) #Tirando o dinheiro da conta do Jorge :(

print(jorge)
````

E como come√ßamos a colocar "as m√£os na massa" iremos entender o porqu√™ do resultado acima, e muito mais al√©m disso.

Ent√£o para come√ßar √© necess√°rio entender que para criar uma classe precisamos utilizar da palavra-chave ```Class``` seguido do nome da nossa classe em quest√£o que geralmente optamos por seguir em Camel Case ou de acordo com a prefer√™ncia do desenvolvedor, mas a priori usamos Camel Case pelo fato das fun√ß√µes utilizarem de Snake Case.

Dando sequ√™ncia, ap√≥s a utiliza√ß√£o da palavra-chave podemos criar atributos pr√©-definidos dentro daquela classe ao qual estes podem vir antes ou depois do construtor tendo cada um sua especificidade exemplificada abaixo:

````python
class Humano:
  dia_de_nascimento = 'Undefined' #Desta forma qualquer modifica√ß√£o neste valor afetar√° todas as inst√¢ncias.
  def __init__(self, nome : str, idade : int, altura : float, peso : float):
    self.nome = nome
    self.idade = idade
    self.altura = altura
    self.peso = peso

  #Podemos referenciar os tipos de dado que cada argumento recebe, al√©m da sa√≠da.

humano_1 = Humano('Jorge', 18, 2.20, 115.7)
humano_1.dia_de_nascimento = '02-01-2026'

humano_2 = Humano('Louise', 19, 1.65, 72)
humano_2.dia_de_nascimento = '31-12-2025'

Humano.dia_de_nascimento = '01-01-2025'
print('Novas inst√¢ncias por padr√£o: ', Humano.dia_de_nascimento)
print('Primeiro humano: ', humano_1.dia_de_nascimento)
print('Segundo humano: ', humano_2.dia_de_nascimento)


# Neste lado temos um atributo de uma classe que por padr√£o quando modificado ir√° modificar o de todas as inst√¢ncias, contudo caso venhamos a atribuir manualmente
# Cada uma delas, por padr√£o a inst√¢ncia modificada tem prioridade e n√£o ser√° modificada mesmo ap√≥s mudan√ßas na classe, contudo a mudan√ßa da classe permanece para todas.
````

## **Abstra√ß√£o**

Na orienta√ß√£o a objeto, a abstra√ß√£o funciona de forma a fragmentar mais e mais os conceitos por tr√°s da ideia a ser representada como classe, de acordo com isso temos duas situa√ß√µes:

* **Aquela que n√£o usamos abstra√ß√£o:** Criei tr√™s classes, uma para representar o Kung Fu, outra para representar o Karat√™, e outra para representar o Muay Thai. Contudo, apesar de serem artes diferentes possuem caracter√≠sticas similares como:
  * Usam os punhos para socos
  * Usam as pernas para chutes

* **Aquela que usamos da abstra√ß√£o:** Criei uma √∫nica classe, que representa os mam√≠feros, nela eu coloco todas as caracter√≠sticas universais que um mam√≠fero precisa ter para que seja ent√£o considerado como um, a partir disso crio classes que herdam essas caracter√≠sticas centrais e dou andamento reutilizando o m√°ximo poss√≠vel do c√≥digo, evitando assim a reescrita de caracter√≠sticas iguais para cada um dos animais mam√≠feros que possuimos.

### **Heran√ßa**

Agora t√° na hora d'eu criar uma classe abstrata e a partir dela utilizar de suas caracter√≠sticas para criar novas classes sem que eu precise realizar a repeti√ß√£o de informa√ß√µes, este processo √© chamado de Heran√ßa e se d√° a partir do momento em que uma classe herda os m√©todos e atributos de uma outra.

````python
class Pessoa: #Super Classe | Classe m√£e, maneiras distintas de chamar.
  def __init__(self, nome, idade):
    self.nome = nome
    self.idade = idade

  def andar(self):
    return f'{self.nome} come√ßou a andar'

  def correr(self):
    return f'{self.nome} come√ßou a correr'

  def sentar(self):
    return f'{self.nome} sentou-se'

class Profissao(Pessoa):
  def __init__(self, nome, idade, profissao, experiencia, salario):
    super().__init__(nome, idade) #Convocamos os atributos da classe m√£e aqui.
    self.profissao = profissao
    self.experiencia = experiencia
    self.salario = salario

  def trabalhar(self):
    return f'{self.nome} come√ßou a empenhar a fun√ß√£o de {self.profissao}'

  def tirar_descanso(self):
    return f'{self.nome} bateu o ponto e foi descansar'

  def ir_embora(self):
    return f'{self.nome} terminou o expediente, bateu o ponto e foi embora'

x = Profissao('Leonardo', 32, 'Confeiteiro', '5 anos', 5200)
print(f'{x.andar()} em dire√ß√£o ao seu trabalho, ao chegar l√° {x.trabalhar()}')

#Lembrando que uma classe m√£e pode dar suas caracter√≠sticas e m√©todos para 'n' classes filhas, e as classes filhas podem fazer o mesmo.
````

````python
#Al√©m de que podemos fazer uma classe filha herdar caracter√≠sticas de duas outras classes, seja m√£e ou n√£o, como por exemplo:

class Corda:
  def __init__(self, tipo_corda):
    self.tipo_corda = tipo_corda

class Instrumento:
  def __init__(self, tipo_do_instrumento):
    self.tipo_do_instrumento = tipo_do_instrumento

class Violao(Corda, Instrumento):
  def __init__(self, material_violao, tipo_corda, tipo_do_instrumento):
    self.material_violao = material_violao
    Corda.__init__(self, tipo_corda)
    Instrumento.__init__(self, tipo_do_instrumento)

  def tocar_musica(self):
    return 'tocando m√∫sica.'

x = Violao('Madeira', 'Nylon', 'Ac√∫stico')
print(f'Meu viol√£o √© feito de {x.material_violao} e tem cordas de {x.tipo_corda} e ele √© {x.tipo_do_instrumento}, agora se me d√° licen√ßa estou {x.tocar_musica()}')
````

### **Encapsulamento**

Na orienta√ß√£o ao objeto, o encapsulamento existe para encapsular os dados de forma a alterar sua acessibilidade, protegendo o c√≥digo de modifica√ß√µes diretas. Dentro do encapsulamento temos tr√™s n√≠veis de acesso s√£o estes:

* P√∫blico
* Protegido
* Particular

No n√≠vel de acesso p√∫blico temos dados que podem ser facilmente acessados e que n√£o possuem restri√ß√µes quanto a sua modifica√ß√£o direta, logo n√£o √© considerado m√° pr√°tica a modifica√ß√£o direta destes atributos caso se fa√ßa necess√°rio. Por√©m, √© importante ter cuidado.

````python
#N√≠vel de acesso - P√∫blico
class Aluno:
  def __init__(self, nome):
    self.nome = nome

jovem = Aluno('Eduardo')
````

````python
print(jovem.nome)
jovem.nome = 'Jos√©'
print(jovem.nome)
````

J√° no n√≠vel de acesso protegido, continuamos sendo capazes de realizar modifica√ß√µes diretas naquele atributo que desejamos, contudo √© de extrema import√¢ncia entender que isso √© considerado uma m√°-pr√°tica e que n√£o √© indicado.

O n√≠vel de acesso protegido n√£o impede a modifica√ß√£o, mas aconselha que estas sejam feitas a partir de m√©todos ou propriedades. Por fim, este n√≠vel de acesso possui como identificador a utiliza√ß√£o de um underline ap√≥s o ponto.

````python
#N√≠vel de acesso - Protegido
class Aluno:
  def __init__(self, nascimento):
    self._nascimento = nascimento

  @property
  def nascimento(self):
    return self._nascimento

  @nascimento.setter
  def nascimento(self, valor):
    self._nascimento = valor
    return f'Data de nascimento alterada para: {valor}'

jovem = Aluno('11-11-2011')
````

````python
jovem._nascimento #Como podemos ver o acesso se d√° a partir da utiliza√ß√£o do underline.
                         #Contudo, caso desejemos seguir as boas pr√°ticas da linguagem, seguimos usando os m√©todos e propriedades para acesso, modifica√ß√£o e exclus√£o.

#Agora vem a parte confusa, do porque criamos dois m√©todos de mesmo nome.
#E isso inclusive na programa√ß√£o orientada a objeto, √© considerado boa pr√°tica se aliado aos decorators. Pois os mesmos realizaram o discernimento sozinhos de qual acionar de acordo com a finalidade, por exemplo:

print(jovem.nascimento)
jovem.nascimento = '25-12-2025'
print(jovem.nascimento)

#Fato interessante: O decorator 'property', √© conhecido por Getter tamb√©m, e tem a funcionalidade que esse nome carrega.
````

Agora trocando o assunto, quando falamos do n√≠vel de acesso part√≠cular estamos nos referindo a um n√≠vel do qual n√£o √© poss√≠vel realizar modifica√ß√µes diretas por conta que a linguagem utiliza de name mangling para impedir que o desenvolvedor acesse diretamente o atributo daquela classe ou m√©todo.


````python
#N√≠vel de acesso - Particular
class Aluno:
  def __init__(self, nome, cpf):
    self.__cpf = cpf
    self.nome = nome

  @property
  def cpf(self):
    try:
      return self.__cpf
    except AttributeError:
      return 'CPF foi deletado e n√£o consta em nosso sistema.'

  @cpf.setter
  def cpf(self, valor):
      self.__cpf = valor

  @cpf.deleter
  def cpf(self):
    del self.__cpf
    print(f'O CPF de {self.nome} foi deletado dos nossos sistemas')

jovem = Aluno('Cloudinho', '123.456.789.10')
print(f'O CPF de {jovem.nome} foi digitado da forma errada e se encontra assim: {jovem.cpf}')

#Ap√≥s isso seguimos com outros exemplos pra colocar em pr√°tica o setter e o deleter.
jovem.cpf = '123.123.456-01' #Setter aplicado
print(f'Sendo assim o CPF de {jovem.nome} agora √©: {jovem.cpf}') #Property aplicado
print(f'A matricula do jovem {jovem.nome} foi cancelada, portanto estamos deletando seu CPF do nosso banco de dados.') #Deleter aplicado ap√≥s este coment√°rio.
del jovem.cpf

print(jovem.cpf)
````

  Contudo ainda vale ressaltar que existem outros decorators que podem ser utilizados, como o Static Method que n√£o √© ligado a classe ou o Class Method que serve como uma f√°brica de objetos mais breve.

### **Polimorfismo**

Em poucas palavras, o polimorfismo √© a capacidade de diversas classes terem a mesma fun√ß√£o com sa√≠das diferentes, como:

````python
class Animal:
  def emitir_som(self):
    raise NotImplementedError

class Gato(Animal):
  def emitir_som(self):
    return f'Miau'

class Cachorro(Animal):
  def emitir_som(self):
    return f'Au au'

x = [Gato(), Cachorro()]
for i in x:
  print(i.emitir_som())
````
