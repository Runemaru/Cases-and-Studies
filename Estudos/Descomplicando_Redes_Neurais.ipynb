## **História da IA**
Tudo começou com a obra seminal "Computing Machinery and Intelligence" publicada em 1950 por Alan Turing que foi o responsável por criar uma pergunta sobre se as máquinas seriam capazes de pensar, contudo foi só após a segunda guerra mundial que então o nome da inteligência artificial começou a se popularizar graças ao John McCarthy também criador da linguagem LISP em 1958. Por fim, ainda em 1958 Frank Rosenblatt constrói o Perceptron I, sendo este o primeiro computador com base em rede neural.

A inteligência artificial é um tópico que foi congelado diversas vezes pelo fato de não obterem avanços científicos, perdendo investimentos e impedindo avanço. O primeiro inverno ocorreu após a limitação exposta por Minsky e Papert documentada no livro "Perceptrons" ao qual demonstrou matemáticamente a incapacidade  do modelo Perceptron de resolver problemas não linearmente separáveis, voltando a reacender os tópicos sobre redes neurais e inteligência artificial apenas após a criação do Backpropagation que falaremos mais a posteriori.

Após isto tivemos o segundo inverno que foi entre os anos de 1980 até meados de 1990, por conta do colapso do mercado de sistemas especialistas, da limitação de hardwares e a dificuldade de escalar estes modelos de redes neurais com os métodos de treino da época. Após isto, retornando apenas após a vinda de múltiplos fatores como: GPUs, Big Data e de avanços algorítmicos como ReLu e  também a criação de técnicas de inicialização.

#### **Características de problemas que podem ser resolvidos por IA.**
* **Natureza Heurística:** Problemas onde o número de soluções cresce exponencialmente, tornando busca exaustiva inviável. Métodos heurísticos exploram seletivamente o espaço, aceitando soluções de alta qualidade sem garantia de otimalidade global.
* **Aprendizado contínuo:** Problemas onde padrões mudam ao longo do tempo, exigindo atualização contínua do modelo sem esquecer conhecimento anterior. Exemplo: Detecção de fraude adaptando-se a novas técnicas, ou sistemas de recomendação ajustando-se a tendências sazonais.
* **Grandes volumes de dados:** Problemas com relações complexas impossíveis de codificar manualmente. Modelos aprendem padrões automaticamente de dados. Grande volume evita overfitting (memorização) e underfitting (simplicidade excessiva). Exemplo: Identificação de tumores via aprendizado supervisionado - CNNs treinadas com milhares de exames rotulados aprendem características discriminativas automaticamente.
Após isto é necessário entender como prosseguir, identificando:

    * O tipo de aprendizagem, se supervisionada, não-supervisionada ou por reforço.
    * E identificar qual a melhor saída, se classificação, regressão, geração ou otimização como exemplo.

## **IA Simbólica & IA Conexionista**
As inteligências artificiais simbólicas foram as primeiras a surgirem quando falamos do desenvolvimento e da história como um todo do tema, dominando as pesquisas entre as décadas de 1950 e 1980. Ela é definida a partir da criação de regras explícitas que se baseiam majoritariamente na matemática e na lógica formal, utilizando representações como lógica de predicados, árvores de decisão e sistemas de regras if-then.

O conhecimento neste paradigma é codificado manualmente por engenheiros de conhecimento, que extraem expertise de especialistas humanos e a transformam em regras processáveis. O processamento ocorre através de mecanismos de inferência lógica, como encadeamento para frente (forward chaining) ou para trás (backward chaining), permitindo que o sistema derive conclusões a partir das regras estabelecidas.

Geralmente este tipo de inteligência possui passos bem direcionados e interpretáveis - é possível rastrear exatamente qual regra foi aplicada em cada decisão. Contudo, são sistemas limitados pelo custo elevado da engenharia de conhecimento (o chamado Knowledge Engineering Bottleneck) e pela rigidez às variações não previstas nas regras, falhando catastroficamente quando confrontados com situações fora do domínio especificado.

Um exemplo clássico de aplicação destas IAs simbólicas é o MYCIN, sistema especialista desenvolvido em Stanford que auxilia em diagnóstico de infecções bacterianas e recomendação de antibióticos, operando com aproximadamente 600 regras de produção e alcançando performance comparável a especialistas humanos em seu domínio específico.

Agora quando falamos da IA Conexionista, estamos adentrando em uma abordagem fundamentalmente diferente. A ideia por trás deste tipo de IA vem do aprendizado a partir de padrões e relacionamentos extraídos dos dados, não de regras programadas explicitamente. O conhecimento aqui é implícito e distribuído nos pesos das conexões de uma rede neural artificial, inspirada no processamento paralelo distribuído observado em sistemas biológicos.

Diferente do paradigma simbólico onde o fluxo é Conhecimento → Regras → Inferência, no conexionismo temos: Dados → Treinamento → Generalização → Validação. O sistema aprende através de ajuste iterativo de parâmetros (pesos sinápticos) utilizando algoritmos como backpropagation e técnicas de otimização, permitindo que o conhecimento emerja do processo de treinamento ao invés de ser explicitamente programado.

A fundamentação teórica para esta capacidade de aprendizado vem do Teorema da Aproximação Universal (Cybenko, 1989; Hornik, 1991), que estabelece que redes neurais com ao menos uma camada oculta podem aproximar qualquer função contínua em domínios compactos (conjuntos fechados e limitados), dado número suficiente de neurônios. Este resultado garante a capacidade representacional das redes, embora não assegure que o treinamento encontrará tal aproximação de forma eficiente — justificando a necessidade de técnicas de otimização e regularização que serão abordadas posteriormente.

Estes sistemas requerem grande volume de dados para treinamento adequado e apresentam um trade-off fundamental: alta performance em tarefas complexas versus baixa interpretabilidade (problema da caixa-preta). Exemplos práticos de sua funcionalidade no mundo real estão presentes no reconhecimento de imagens através de CNNs (Convolutional Neural Networks), processamento de linguagem natural com Transformers, e demais arquiteturas de Deep Learning - que representa a evolução moderna do paradigma conexionista com redes de múltiplas camadas capazes de aprender representações hierárquicas complexas.

## **Perceptron: Estrutura e Limitações**

O Perceptron como entendemos em um dos primeiros capítulos foi uma criação de Frank Rosenblatt, ao qual foi o primeiro algoritmo a simular um neurônio biológico com mais exatidão, contudo este enfrentou limitações das quais iremos discorrer ao longo deste tema. Além disto abaixo temos a fórmula matemática por trás do Perceptron:

$$y = f\left(\sum_{i=1}^{n} x_i \cdot w_i + b\right)$$

Nela temos o entendimento que queremos encontrar $y$, mas que para encontrar este valor utilizamos da função $f\left(\sum_{i=1}^{n} x_i \cdot w_i + b\right)$ que é basicamente a somatória dos produtos obtidos pela multiplicação de $x_i$ com $w_i$ que são respectivamente as entradas e pesos em questão, para só então após a realização da somatória obtermos o valor a ser somado com o Bias representado pelo $b$ e que nós definimos durante o treino.

* **Pesos:** Controlam a inclinação da nossa reta
* **Bias:** Controlam a fronteira de decisão por onde passsa a nossa reta.

O Perceptron mais simplitas por conta de suas limitações sempre irá criar fronteiras específicas de acordo com o espaço em que está trabalhando, dentre os tipos de fronteiras temos:

* Linhas retas, para ambientes 2D (Duas entradas)
* Plano, para entradas em ambiente 3D (Três entradas)
* Hiperplano, para ambientes acima do 3D ($n$ entradas)

Além disto, conforme as iterações se passam o Perceptron reavalia os seus pesos e bias de acordo com o resultado obtido utilizando das fórmulas:
* **Para obter novos pesos:** $\text{Peso novo} = \text{Peso atual} + (\eta * e * x_i)$
* **Para obter novo bias:** $\text{Bias novo} = \text{Bias atual} + (\eta * e)$



````python
#Antes de começarmos a criar o Perceptron, é necessário entender que o mesmo não aceita matrizes.
#Seus pesos podem ser definidos de maneira aleatória ou podem ser todos zeros, tudo dependerá da dinâmica em questão.
#Os pesos, bias são atualizados durante o processo de execução do código.

import numpy as np
np.random.seed(6)

#Criação das amostras para representação no Perceptron.
matriz_completa = np.random.randint(0,3, (10, 3))
resposta_esperada = [1, 1, 0, 1, 1, 1, 0, 0, 1, 1]

#Divisão entre conjuntos de treino e teste.
matriz_treino = matriz_completa[:7]
resposta_treino = resposta_esperada[:7]

matriz_teste = matriz_completa[7:]
resposta_teste = resposta_esperada[7:]
print(f'Linhas e colunas de teste: {matriz_teste.shape}') #Contêm 30% do conjunto.
print(f'Linhas e colunas de treino: {matriz_treino.shape}') #Contêm 70% do conjunto.
print('-' * 35)

matriz_completa
````

Após realizarmos a divisão entre os conjuntos de treino e teste iremos aplicar os conceitos do Perceptron contruiremos o modelo via POO.

````python
class Perceptron():
  def __init__(self, amostra, taxa_aprendizagem = 0.1, epoca = 5): #Setei valores padrões nos hiperparâmetros e que podem ser ajustados caso o usuário deseje.
    self.amostra = amostra
    self.taxa_aprendizagem = taxa_aprendizagem
    self.epoca = epoca
    self.bias = 0
    self.peso = np.zeros(amostra.shape[1]) #Os pesos podem ser pré-determinados (via aleatorização de um espaço pequeno Ex: -0.5 e 0.5) ou começar com zero
                                      #os pesos também passam por modificações conforme as iterações dentro do próprio Perceptron como veremos adiante.

  def sumario(self, amostra):
    return print(f'''\tSumário do nosso modelo
    Tamanho amostra: {len(amostra)}
    Taxa de Aprendizagem: {self.taxa_aprendizagem}
    Número de épocas: {self.epoca}
    Bias atual: {self.bias}
    Peso atual: {self.peso}''')

  def funcao_ativacao(self, soma):
    if soma > 0: return 1
    else: return 0

  def calcular_saida(self, entrada):
    soma = np.dot(entrada, self.peso) + self.bias
    saida = self.funcao_ativacao(soma)
    return saida

  def treinar_modelo(self, entrada, resposta):
    for treinos in range(1, self.epoca + 1):
      erros_na_epoca = 0

      print('-' * 60)
      print(f'ÉPOCA {treinos}')
      print('-' * 60)

      for i in range(len(entrada)):
        posicao_amostra = entrada[i]
        resposta_correta = resposta[i]

        saida = self.calcular_saida(posicao_amostra)
        erro = resposta_correta - saida

        self.peso = self.peso + (self.taxa_aprendizagem * erro * posicao_amostra)
        self.bias = self.bias + (self.taxa_aprendizagem * erro)

        if erro != 0:
          erros_na_epoca += 1
          print(f'Amostra nº{i}: Margem de erro {erro} | Bias → [{self.bias}] | Peso → [{self.peso}].')
        else: print(f'Amostra nº{i}: Margem de erro {erro}')

      print(f'\n Época {treinos}, teve um total de {erros_na_epoca} erros entre as {len(entrada)} amostras')
      if erros_na_epoca == 0:
        print(f'Modelo convergiu, encerrando na ÉPOCA {treinos}')
        break


  def prever_resultados(self, entrada):
    resultados = [self.calcular_saida(vetores) for vetores in entrada]
    return resultados

````

Após criarmos o nosso modelo Perceptron, precisamos voltar para fixar ainda mais alguns conceitos que são fundamentais para que haja o entendimento real de como essa rede neural funciona, são elas:

* Os hiperparâmetros são modificados até encontrar uma combinação que consiga acertar todas as amostras, logo, não existe hiperparâmetro para cada amostra.
* Se duas amostras são idênticas e possuem rótulos diferentes, as mesmas impedem do modelo convergir, trazendo um tipo de não-linearidade.
* Se a estrutura que contêm as amostras é inerentemente não-linear, o modelo não será capaz de trabalhar com essas estruturas por conta de sua própria limitação, este problema ficou conhecido como XOR, ou **eXclusive OR**, sendo este problema uma operação lógica binária que retorna 1 quando as entradas são diferentes e 0 quando as entradas são iguais.

## **Funções de ativação**

As funções de ativação são de extrema importância para o nosso modelo, pois são elas que introduzem a não-linearidade ao mesmo, pois caso estas não fossem aplicadas tudo o que teriamos seria uma reta.

Neste tópico obteremos uma base concisa e que nos permita ter um entendimento pleno sobre o que se diz das funções de ativação em geral. E para que consigamos entender o impacto direto das funções de ativação, disponibilizarei este vídeo que mostra a diferença entre três funções de ativação na prática.

* **[Sobre o vídeo:](https://https://www.youtube.com/shorts/xCDtBsqMFpo)** Durante o percorrer do vídeo podemos perceber que cada função de ativação teve um comportamento diferente ao longo das épocas passadas. É importante entender que apesar de adicionarem não-linearidade, nem sempre a função de ativação a ser utilizada é a melhor, logo é importante sempre buscar melhorar o desempenho através de outras que nos sirvam melhor. E é claro que reforçando o conhecimento, sem a função de ativação teriamos apenas uma reta.

### **Função de ativação: Step (Função Degrau)**

Essa função de ativação tem sua origem no modelo Perceptron, logo sua fórmula é a mesma da qual tivemos conhecimento no modelo que construimos a priori, sendo ela:

$$f(x) = if(x) \geq 0 : 1 \text{ else } 0$$


Nesta função de ativação temos algumas características como:
- **Comportamento binário:** Só existem saídas iguais a 0 ou 1.
- **Descontinuidade:** A função traz um resultado descontinuo, pois por não existir transição suave do 0 ao 1, a mesma acaba que por saltar instantaneamente para o valor.
- **Problema da derivada:**  Essa função de ativação possui derivada sempre igual a zero, pois sua função tem resultado constante.
  - **Consequências:** Por conta da derivada ser sempre igual a zero, o algorítmo de Backpropagatio não consegue realizar o cálculo de gradientes multiplicando as derivadas em cadeia, pois se uma derivada é igual a zero, todo o produto é zero. Por conta disto, os pesos não recebem informação de como ajustar o problema em questão.

E foram por estas complicações que logo a função de ativação se tornou obsoleta ao passar dos anos, sendo substituida ao passar dos anos por outras funções de ativação que conseguiam trazer resultados melhores. Abaixo crio um exemplo de como seria a visualização da função de ativação em questão para termos uma melhor dinâmica.

````python
import numpy as np
import matplotlib.pyplot as plt

# Definição da função Step
def step(x):
    return np.where(x >= 0, 1, 0)

x = np.linspace(-10, 10, 1000)

y = step(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b-', linewidth=2)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Função de Ativação Step')
plt.ylim(-0.1, 1.1)
plt.show()
````

### **Função de ativação: Sigmoid (Logística)**

A função sigmoide tem como objetivo a classificação binária dos valores em questão, e é muito utilizada quando queremos usar valores que estão entre 0 e 1 sem nunca alcancá-los. Sendo o intervalo aberto entre (0, 1).
$$\sigma(x) = \frac{1}{1 + e^{-x}}$$
$$\text{1 dividido por 1 somado a base de euler ao expoente de -x resulta em sigma de X}$$

$$\text{Enquanto a sua derivada é denotada por:}$$

$$\sigma'(x) = \sigma(x) \cdot (1 - \sigma(x))$$

É de grande ajuda entender que a derivada da função de ativação Sigmoid utiliza do resultado da função de ativação para calcular sua derivada, que será utilizada para o reajuste dos pesos durante o Backpropagation que será explicado mais tarde.

````python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def derivada_sigmoid(x):
  return sigmoid(x) * (1 - sigmoid(x))

x = np.linspace(-10, 10, 1000)
y = sigmoid(x)

x = np.linspace(-10, 10, 1000)
y = sigmoid(x)
y_deriv = derivada_sigmoid(x)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(x, y, 'b-', linewidth=2)
axes[0].axhline(y=0, color='k', linewidth=0.5)
axes[0].axhline(y=1, color='r', linestyle='--', alpha=0.5)
axes[0].axhline(y=0, color='r', linestyle='--', alpha=0.5)
axes[0].axvline(x=0, color='k', linewidth=0.5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('Sigmoid(x)')
axes[0].set_title('Função Sigmoid')
axes[0].grid(True, alpha=0.3)

axes[1].plot(x, y_deriv, 'r-', linewidth=2)
axes[1].axhline(y=0.25, color='g', linestyle='--', alpha=0.5, label='Máximo = 0.25')
axes[1].axvline(x=0, color='k', linewidth=0.5)
axes[1].set_xlabel('x')
axes[1].set_ylabel("Sigmoid'(x)")
axes[1].set_title('Derivada da Sigmoid')
axes[1].grid(True, alpha=0.3)
axes[1].legend()

plt.tight_layout()
plt.show()
````

Contudo, é interessante trazer que por limitações computacionais de precisão, quando chegamos a valores extremos os computadores tendem a arredondar para 0 ou 1, mesmo que na lógica matemática não consigamos chegar a 1 ou 0 por conta da propriedade da fórmula em questão.

Agora quando falamos sobre limitações do Sigmoid temos:

1. **Vanishing Gradient (Gradiente Desvanecente):** Quando realizamos o treinamento da nossa rede neural, o gradiente (informação usada para ajustar os nossos pesos) precisa viajar da última até a primeira camada. Em cada camada, o gradiente é multiplicado pela derivada da função de ativação. Como a derivada máxima da sigmoid é 0.25, o gradiente é reduzido a cada camada que atravessa. Conforme ele vai voltando pelas camadas, o valor acaba por se reduzir tanto ao ponto de praticamente não causar mudanças relevantes nos pesos das primeiras camadas.
2. **Saídas não centradas em zero:** A sigmoid produz valores entre 0 e 1, portanto todas as saídas são positivas. Quando um neurônio recebe entradas de neurônios anteriores que usam sigmoid, todas essas entradas são positivas. Durante o backpropagation, isso faz com que os gradientes de todos os pesos desse neurônio tenham o mesmo sinal. Consequentemente, ou todos os pesos aumentam juntos, ou todos diminuem juntos. Não é possível que um peso aumente enquanto outro diminui na mesma iteração, forçando a otimização a seguir um caminho em zig-zag ao invés de convergir diretamente ao ponto ótimo.
3. **Saturação:** Na função sigmoid temos a problemática da saturação que ocorre devido a valores que são muito distantes de 0 tanto positivamente quanto negativamente. A saturação impacta diretamente no modelo em que estamos treinando, e faz com que o gradiente resultante de nossa função de ativação se torne insignificante, os pesos praticamente não sejam atualizados e faça com que o neurônio que opera em questão pare de aprender efetivamente. Além disto, temos três áreas dentro do gráfico que denominamos como:
  * Saturação superior: Intervalo de dados positivos distantes de 0.
  * Área ativa: Intervalo de dados mais próximos de 0 e que permitem visualização de mudança no gráfico em questão como podemos ver no gráfico gerado acima.
  * Saturação inferior: Intervalo de dados negativos distantes de 0.

### **Função de ativação: Tanh (Tangente Hiperbólica)**

A função Tanh transforma qualquer valor em um valor entre -1 e 1, e veio como uma solução ao problema de saídas não centradas em zero da função de ativação Sigmoid, graças a esta funcionalidade a função Tanh consegue uma convergência melhor do neurônio graças a possibilidade dos gradientes terem sinais diferentes. Sua fórmula é denotada por:
$$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$
$$\text{Enquanto a sua derivada é denotada por:}$$
$$\tanh'(x) = 1 - \tanh^2(x)$$

A função em questão foi muito popular antes do surgimento da ReLU e ainda é utilizada em algumas arquiteturas, principalmente em redes recorrentes (RNNs e LSTMs), essencialmente a Tanh é uma sigmoid reescalada e deslocada, onde para esticar a curva vertical multiplicamos por 2 e para mover a curva mais para baixo subtraimos 1, obtendo assim $\tanh(x) = 2 \cdot \sigma(2x) - 1$



Abaixo temos como a representação gráfica do Tanh é realizada:

````python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

def tanh_derivada(x):
    return 1 - np.tanh(x)**2

x = np.linspace(-10, 10, 1000)
y = tanh(x)
y_deriv = tanh_derivada(x)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(x, y, 'b-', linewidth=2)
axes[0].axhline(y=0, color='k', linewidth=0.5)
axes[0].axhline(y=1, color='r', linestyle='--', alpha=0.5)
axes[0].axhline(y=-1, color='r', linestyle='--', alpha=0.5)
axes[0].axvline(x=0, color='k', linewidth=0.5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('tanh(x)')
axes[0].set_title('Função Tanh')
axes[0].grid(True, alpha=0.3)

axes[1].plot(x, y_deriv, 'r-', linewidth=2)
axes[1].axhline(y=1, color='g', linestyle='--', alpha=0.5, label='Máximo = 1')
axes[1].axvline(x=0, color='k', linewidth=0.5)
axes[1].set_xlabel('x')
axes[1].set_ylabel("tanh'(x)")
axes[1].set_title('Derivada da Tanh')
axes[1].grid(True, alpha=0.3)
axes[1].legend()

plt.tight_layout()
plt.show()
````

Dentre então suas vantagens e desvantagens temos:

* **Correção do problema de saída não centrada em zero:** Comparado a função Sigmoid, a função de ativação Tanh solucionou o problema sendo centrada entre -1 e 1 e consequentemente aumentado o alcance de sua derivada.
* **Derivada máxima maior:** A derivada máxima de Sigmoid é 0.25, contudo a de Tanh é quatro vezes maior (1.0) reduzindo o problema de Vanishing Gradient.
* **Função de ativação simétrica:** A função Tanh é simétrica, logo $\tanh(-x) = -\tanh(x)$

Enquanto dentre as desvantagens temos algumas já conhecidas e outras que iremos discorrer um pouco mais sobre, como:

* **Vanishing Gradient:** Apesar de reduzir o problema, naõ extingue, logo valores distantes de zero continuam impactando diretamente o treinamento do nosso modelo. Contudo, deve-se levar em consideração que no caso de modelos sem muitas camadas o problema tem menor impacto.

* **Saturação:** Assim como Sigmoid, a Tanh satura quando distante de zero.

* **Custo computacional:** Como a função requer cálculo de duas exponênciais e realiza operações de divisão e subtração tende a ser mais custosa computacionalmente que funções mais simples como ReLu.

Por fim, é interessante trazer algumas propriedades desta função de ativação em questão, como por exemplo:

* **Intervalo matemático é (-1, 1), ou seja aberto e há exclusão**
* **Valor em X = 0, este é o ponto central da função onde a derivada é máxima.**

## **Função de ativação: ReLu (Rectified Linear Unit)**

A função ReLu se tornou a função de ativação mais utilizada em redes neurais modernas devido a sua simplicidade, baixo custo computacional e pelo fato de resolver problemas como Vanishing Gradient. A ReLu é denotada por: $f(x) \max(0, x)$ enquanto a sua derivada trabalha de forma condicional onde se $x > 0$ o resultado será sempre $1$, e se $x \leq 0$ resultado sempre será $0$.

Abaixo podemos ver na prática como a função de ativação ReLu se comporta na prática e entender sua projeção.


````python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def relu_derivada(x):
    return np.where(x > 0, 1, 0)

x = np.linspace(-5, 5, 1000)
y = relu(x)
y_deriv = relu_derivada(x)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico da função
axes[0].plot(x, y, 'b-', linewidth=2)
axes[0].axhline(y=0, color='k', linewidth=0.5)
axes[0].axvline(x=0, color='k', linewidth=0.5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('ReLU(x)')
axes[0].set_title('Função ReLU')
axes[0].grid(True, alpha=0.3)
axes[0].set_ylim(-1, 5)

# Gráfico da derivada
axes[1].plot(x, y_deriv, 'r-', linewidth=2)
axes[1].axhline(y=1, color='g', linestyle='--', alpha=0.5, label='Derivada = 1')
axes[1].axvline(x=0, color='k', linewidth=0.5)
axes[1].set_xlabel('x')
axes[1].set_ylabel("ReLU'(x)")
axes[1].set_title('Derivada da ReLU')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].set_ylim(-0.5, 1.5)

plt.tight_layout()
plt.show()
````

Dentre suas vantagens temos:

  * **Resolução do Vanishing Gradient:** Sua derivada é constante e igual a $1$ para qualquer $x > 0$, logo não importa quão grande é $x$.

  * **Custo computacional baixo:** ReLu é uma função de ativação simples e que executa com baixissimo custo computacional se comparada a funções como Tanh.

  * **Não satura na região positiva:** Valores distantes de zero positivamente não saturam nesta função de ativação linear, mantendo a reta constante.

  * **Esparsidade:** ReLu desliga neurônios negativos transformando-os em zero tornando o conjunto de dados esparso, isso pode ser tanto uma vantagem quanto uma desvantagem que indica Dying ReLu, mas dentre suas vantagens temos:
    * Generalização melhorada.
    * Redução de overfitting
    * Aumento da eficiência computacional.

Contudo, após entendermos suas vantagens, entenderemos que as desvantagens por traz desta função de ativação:

  * **Dying ReLU:** Se o neurônio consistentemente receber entradas que são $leq 0$ o mesmo desligará os neurônios, e isso ocorre pelo fato de que quando ReLu determina o valor de entrada como zero, todos os seus valores subsequentes serão zero e entãoos pesos não serão atualizados. Dentre as causas comuns de ocorrência temos:
    * Learning rate muito alto.
    * Inicialização ruim de pesos.
    * Bias muito negativo

  * **Não centrada em zero:** Por só produzir valores não-negativos acaba por se tornar não centrada em zero e consequentemente ocasionar em atualizações zig-zag, impedindo a convergência rápida do modelo, comportamento igual a Sigmoid.
  * **Saturação inferior:** Existe possibilidade de saturação inferior desde que os valores permaneçam em zero.
  * **Não é diferenciável de $x = 0$:** A função possui uma "quina" em x = 0, onde a inclinação muda abruptamente de 0 para 1. Matematicamente, isso significa que não existe uma única reta tangente nesse ponto, portanto a derivada não está definida ali. Apesar disto, raramente causa problemas pelo fato de que:
    * A probabilidade de $x$ ser exatamente 0 é muito baixa.
    * E por convenção, define-se $f'(0) = 0$
  * **Não é limitada superiormente:** Pode receber valores de $0$ a $∞^+$ o que pode causar problemas como explosão de ativações, instabilidade numérica e necessidade de normalização.

### **Leaky ReLU (A solução contra os Dying Neurons)**

Esta função de ativação foi criada no intuito de impedir que houvessem neurônios mortos durante a aplicação da função em questão. E ela vem com uma mudança símbolica na forma que é tratada a condicional do ReLU. Sendo sua denotação matemática:

$f(x) = x$ se $x > 0$ ou caso $x \leq 0$ então $\alpha \cdot x$

O que permite o neurônio não morrer e atualizar seus pesos, mesmo que de forma lenta e gradual, abaixo temos seu comportamento.

````python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def leaky_relu_derivada(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

x = np.linspace(-5, 5, 1000)
y_relu = relu(x)
y_leaky = leaky_relu(x)
y_deriv = leaky_relu_derivada(x)

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Gráfico 1: Comparação ReLU vs Leaky ReLU
axes[0].plot(x, y_relu, 'b-', linewidth=2, label='ReLU')
axes[0].plot(x, y_leaky, 'r-', linewidth=2, label='Leaky ReLU (α=0.01)')
axes[0].axhline(y=0, color='k', linewidth=0.5)
axes[0].axvline(x=0, color='k', linewidth=0.5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('f(x)')
axes[0].set_title('ReLU vs Leaky ReLU')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Gráfico 2: Leaky ReLU ampliada (para ver a inclinação negativa)
axes[1].plot(x, y_leaky, 'r-', linewidth=2, label='Leaky ReLU')
axes[1].axhline(y=0, color='k', linewidth=0.5)
axes[1].axvline(x=0, color='k', linewidth=0.5)
axes[1].set_xlabel('x')
axes[1].set_ylabel('f(x)')
axes[1].set_title('Leaky ReLU (observe a pequena inclinação para x < 0)')
axes[1].legend()
axes[1].grid(True, alpha=0.3)
axes[1].set_ylim(-0.5, 5)

# Gráfico 3: Derivada
axes[2].plot(x, y_deriv, 'g-', linewidth=2, label='Derivada')
axes[2].axhline(y=1, color='r', linestyle='--', alpha=0.5, label='Derivada = 1 (x > 0)')
axes[2].axhline(y=0.01, color='b', linestyle='--', alpha=0.5, label='Derivada = 0.01 (x ≤ 0)')
axes[2].axvline(x=0, color='k', linewidth=0.5)
axes[2].set_xlabel('x')
axes[2].set_ylabel("f'(x)")
axes[2].set_title('Derivada da Leaky ReLU')
axes[2].legend()
axes[2].grid(True, alpha=0.3)
axes[2].set_ylim(-0.1, 1.2)

plt.tight_layout()
plt.show()
````

Dentre suas vantagens temos:

* **Resolução do Dying Neurons**
* **Continuamos com todas as vantagens do ReLU**
* **Permite valores negativos em sua saída.**

Já entre suas desvantagens temos:

* **O valor de $\alpha$ é um hiperparâmetro.**
* **Não temos garantia de qual valor de $\alpha$ é melhor para o modelo.**
* **Ainda não é diferenciável de 0.**
* **Não é limitada superiormente.**

### **Arquitetura de um Perceptron Multicamadas, como funciona?**

Como podemos ver na imagem acima, o Perceptron Multicamadas ou MLP, funciona através da adição de uma nova camada que conhecemos como Hidden Layer ou Camada Oculta em nossa lingua materna. É interessante citar que podemos ter quantas camadas ocultas forem necessárias para o problema ao qual estamos tentando resolver.

Agora falando sobre cada uma destas camadas, temos:

* **Input Layer (Camada de entrada):** Na camada de entrada, realizamos a inserção dos nossos dados.

### **Como o MLP resolveu o problema XOR?**

O problema XOR foi resolvido graças a inserção de camadas ocultas que seriam ditos como nossos neurônios artíficiais. O que acontece para a resolução do problema XOR é que cada neurônio calcula com seus próprios pesos valores e obtem retas que separam cada um dos pontos em questão, após cada um traçar uma reta a saída cria uma junção destas duas retas obtendo o resultado e criando a separação dentre os elementos presentes.

## **Função de Custo (Loss Function)**

As funções de custo/perda tem como objetivo determinar o quanto uma determinada rede neural falhou em classificar ou prever números, ou seja, existem funções tanto para modelos de regressão quanto para modelos de classificação.

## **Backpropagation**

Para que entendamos Backpropagation creio que é ideal entendermos o fluxo como um todo do treinamento das redes neurais, então vamos começar por partes:
1. O modelo recebe os dados de entrada (Input Layer)
2. O modelo segue para as camadas ocultas e passa por todos os neurônios de cada uma das camadas, tendo assim seu resultado modificado. (Hidden Layer)
3. O modelo chega a camada de saída e passa pelo último cálculo de somatória e execução de sua função de ativação em questão.
4. Coletamos o resultado da camada de saída e comparamos com a saída esperada, aplicando uma função de custo/perda para avaliar o quão distante o modelo está do resultado esperado.
5. O Backpropagation entra em ação coletando o valor do erro e calculando o seu gradiente, a partir disto voltará por todas as camadas até a entrada orientando cada uma destas do seu erro e permitindo que as mesmas ajustem os seus pesos.

É importante ressaltar que o Backpropagation é apenas o processo de volta dos dados, e não realiza o processo que o Feedforward realiza, logo o que acontece é que no processo de treinamento tudo fica conectado e interage entre si para otimização geral.


## **Otimizadores**

Enquanto o Backpropagation volta relatando o quanto aquele modelo errou, temos os Otimizadores que farão o trabalho de atualizar os nossos pesos para que consigamos chegar no menor erro possível.

Só que para entendermos os otimizadores precisamos seguir de forma muito mais aprofundada, começando pelo que chamamos de Descent Gradient que é a base por trás de muitos otimizadores, a fórmula se denota por:
$$W_{novo} = W - \eta \cdot \nabla J(W)$$

Onde cada simbolo representa:
* $W$ : Os pesos em questão.
* $\eta$ : O taxa de aprendizagem
* $\nabla J(W)$ : O valor do Gradiente obtido pelo Backpropagation a partir do erro do peso em questão.

Contudo, temos um agravante sobre a fórmula acima, a mesma não ganha velocidade adequada e foi a partir desta que criaram novos otimizadores visando aprimorar o fluxo de diversos meios de aplicação das redes. Dentre os otimizadores podemos citar exemplos como:

* **SGD (Stochastic Gradient Descent):** Seria a fórmula exibida acima, contudo ele acaba por oscilar muito pois ele coleta uma amostra única e calcula o erro apenas para aquela amostra replicando para todas as outras o peso em questão. E por conta disto tende a demorar de convergir e pode se enganar com o que podemos chamar de mínimos locais, onde basicamente temos valores baixos que convergem, porém não são o esperado, e como a fórmula só permite descida o reajuste fica impedido de dar continuidade.
* **SGD com Momentum (O Físico):** Vem como solução ao SGD, introduzindo um conceito da física chamado Momento Linear (Inércia). O problema em questão é resolvido com a adição de uma nova fórmula.
$$V_{novo} = (\beta \cdot V_{antigo}) + ((1 - \beta) \cdot \nabla J)$$
Onde temos em cada incógnita:
    * $V$ : Velocidade começa como 0 e a fórmual é aplicada desde o início.
    * $\beta$ : Geralmente por padrão é 0.9, servindo como um regulador de ordem para o Gradiente, evitando que hajam mudanças bruscas em valor.
    * $\nabla J(W)$ : Gradiente em função do peso atual.
    * Por fim, após passarmos pela formulação inicial, substituimos o $\eta$ pela velocidade $V$ no cálculo de reajuste de pesos, tendo a seguinte fórmula:

    $$W_{novo} = W_{atual} - (\eta \cdot V_{novo})$$

* **Adam:** O Adam diferente dos otimizadores apresentados anteriormente além de ser considerado por muitos o melhor otimizador, temos uma complexidade maior também para o entendimento do mesmo, pois ele executa duas funções ao mesmo tempo para otimizar os nossos pesos.

Para começar com a explicação deste otimizador em questão precisamos entender que antes de iniciar o treinamento o mesmo tem valores padrões, são estes que aparecerão a posteriori:

* $m_0 :$ Conhecido como primeiro momento, é a variável que irá acumular a média histórica da direção dos gradientes, ele inicia por padrão com zero pois o treinamento não se iniciou, assim como os próximos.
* $v_0 :$ Conhecido como segundo momento, acumula a média histórica da  magnitude (intensidade) dos gradientes.
* $t :$ É o contador tempo, sempre que atualizarmos o peso, ele aumentará.

---

Após entendermos o stage inicial do otimizador Adam, seguimos para o momento em que o treinamento se deu início. Então como primeiro passo realizamos o cálculo do gradiente atual que será representado no Adam como $g_t$.
$$g_t = \nabla J(W_{t-1})$$

O que acontece na fórmula que vemos acima é basicamente a subtração entre o peso atual e o produto de $\eta$ pelo nosso gradiente em função do peso anterior a este momento.

---

Após esta etapa o Adam irá atualizar o primeiro momento denotado por $m_t$, então o que iremos obter para esse momento em questão é a média móvel da direção. Com isto será necessário utilizar da fórmula a seguir:


$$m_t = \beta_1 \cdot m_{t-1} + (1 - \beta_1) \cdot g_t$$

* **$m_t$ :** A média móvel da direção atual.
* **$m_{t-1}$ :** A média móvel da direção anterior.
* **$\beta_1$ :** É um hiperparâmetro que por padrão é 0.9 e tem a mesma funcionalidade que o SGD Momentum.
* **$(1 - \beta_1)$ :** Traz o valor percentual do quanto o peso impactará diretamente na mudança do nosso valor em questão.
* **$g_t$ :** Esta incógnita serve como um filtro suavizador. Ruídos são amortecidos pela inércia em $m_{t-1}$
---

Depois que atualizamos o primeiro momento seguimos para o segundo momento denotado por $v_t$.

$$v_t = \beta_2 \cdot v_{t-1} + (1 - \beta_2) \cdot g_t^2$$

* **$v_t$ :** Valor atual da magnitude
* **$\beta_2$ :** Hiperparâmetro que por padrão neste passo será 0.999, pois o Adam precisa de um histórico muito longo da magnitude para ser estável.
* **$g_t^2$ :** O gradiente ao quadrado elimina o sinal negativo e mede a volatilidade ou energia do gradiente, não sua direção.

---

Então chegamos a correção de viés denotada por $\hat{m}$ e $\hat{v}$

A função principal desta etapa é corrigir o problema matemático de termos iniciado com valores em zero para a nossa média móvel e para nossa magnitude, com isso vamos precisar inflar os valores apenas no começo a partir da seguinte fórmula destrinchada a seguir:
$$\hat{m}_t = \frac{m_t}{1 - \beta_1^t} \quad \text{e} \quad \hat{v}_t = \frac{v_t}{1 - \beta_2^t}$$

Logo com base nos resultados de $m_t$ e $v_t$, temos os outros parâmetros presentes que terão a seguinte ideia:

* **$
\hat{m}_t$ e $\hat{v}_t$ :** Lemos como M-Chapéu e V-Chapéu, e são os nossos valores desenviesados.
* **$\beta_1^t$:** Este é o Beta elevado à potência do tempo atual, logo se essa é a terceira rodada do fluxo em questão do Adam, temos que $\beta_1^3$
---
Após finalmente termos desenviesado os nossos dados seguimos para a atualização dos pesos, sendo este o último passo que o otimizador percorre. Sua fórmula final será denotada por:
$$W_t = W_{t-1} - \eta \cdot \frac{\hat{m}_t}{\sqrt{\hat{v}_t} + \epsilon}$$

* **$W_t$ :** Peso atualizado.
* **$\eta$ :** Taxa de aprendizagem.
* **$\sqrt{\hat{v}_t}$ :** A raiz quadrada de v-chapéu traz a magnitude que estava ao quadrado de volta para a escala original, neste caso linear.
* **$\epsilon$ :** Um número minúsculo que é adicionado para trazer estabilidade numérica e impedir que haja divisão por zero.

````python
#Importação de bibliotecas.
import tensorflow as tf
from tensorflow.keras import layers, models
````

````python
#Aqui iremos importar e já separar os dados de treino e teste.

(X_treino, y_treino), (X_teste, y_teste) = tf.keras.datasets.mnist.load_data()
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_treino.shape
````

Após importarmos as bibliotecas que forem necessárias estudamos o dado ou o conhecemos a priori para entender que precisaremos realizar $n$ modificações das quais auxiliarão nossas redes neurais a terem uma atuação relevante na nossa predição.

Para o MNIST precisamos saber que os dados:
* Variam de 0 a 255
* Cada um desses valores $n$ representa um número que vai de $0$ a $9$.
* O dado possui duas dimensões, como vimos no shape, são estas altura e largura.

E para solucionar esse tipo de problema precisamos normalizar os dados e achatá-los para que os mesmos comportem o formato aceito pela nossa rede neural em questão que seria $n \times m$

````python
#Normalização dos valores.

X_treino, X_teste = X_treino / 255.0, X_teste / 255.0
````

````python
#Criação da arquitetura da rede.

modelo = models.Sequential([
    layers.Flatten(input_shape = (28, 28)), #Aqui acontece o achatamento do dado pela multiplicação da altura * largura.
    layers.Dense(128, activation = 'relu'), #Criação da nossa camada oculta, definimos a função de ativação e a quantia de neurônios.
    layers.Dense(10, activation = 'softmax') #Aqui estamos modulando nossa camada de saída com a quantia de neurônios sendo o total de números.
                                                #O softmax é uma função de ativação usada muito pra classificação multiclasse.

]) #A partir disto finalizamos a criação de nossa arquitetura inicial, contudo como vamos treinar o nosso modelo ainda é muito importante
   #Seguirmos para o processo de compilação do nosso modelo em questão, que será onde iremos configurar
   #As nossas métricas de avaliação, otimizadores de peso e nossa função de perda a ser utilizada.
````

````python
#Configurando nossos otimizadores de pesos, funções de perda e métricas de avaliação.

modelo.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
````

Ainda não inicializamos os pesos, pois no próprio Tensorflow temos por padrão o parâmetro ```glorot_uniform``` que será utilizado apenas para ilustrar, pois para a função de ativação ```Relu```temos como melhor inicializador o ```he_normal```. e apenas como disclaimer para escolhermos qual será o nosso inicializador de pesos, podemos utilizar do hiperparâmetro ```kernel_initializer``` enquanto para o nosso viés temos o ```bias_initializer``` que por padrão é $0$.

Logo, dando sequência, após compilarmos o nosso modelo podemos iniciar o seu treinamento direto para ver se tivemos bons resultados.

````python
modelo.fit(
    X_treino,
    y_treino,
    batch_size = 30,
    epochs = 3
)
````

Para a função ```fit()```temos de trazer o valor ao qual iremos treinar $X$ e suas respostas $y$, além disso também utilizamos de dois hiperparâmetros.

* **batch_size:** Quantia de amostras que o modelo irá visualizar antes de ajustar os seus pesos.
* **epochs:** Quantia de vezes que o modelo vai percorrer o dataframe inteiro durante o treinamento.

````python
predicao_teste = modelo.predict(X_teste)
````

````python
modelo.summary() #Com este método podemos verificar quanto cada camada irá precisar treinar dentro do nosso modelo.
#Lembrando sempre que cada neurônio recebe as 784 linhas, logo temos 128 neuronios cada um para 784 linhas.
#Equivalendo em 100.352, e por fim os viéses que são 1 por neurônio.
#Após a camada oculta, passamos pra última camada que sempre será lida como camada de saída.
#Nela teremos os 128 resultados obtidos de cada neurônio multiplicado por 10 neurônios, o que equivale a 1280 saídas + 1 viés por neurônio.
````

Após então treinarmos nosso modelo, podemos seguir para sua predição e validar o resultado por trás do mesmo. É importante entender que a função de ativação ```softmax``` traz uma matriz de $10 \times 10$ representando o percentual que cada valor dentre aqueles presentes no MNIST tem de ser de fato aquele valor. Em outras palavras, o quanto a entrada $n$ pode ser $0$, $1$, $2$, $3$, $n$...

````python
#Agora que tenho o meu modelo treinado, e que já previ os valores em questão.
#Como consigo gerar algo mais visual pra entender o quanto aquele dado acertou?

from sklearn.metrics import confusion_matrix
confusion_matrix(predicao_teste, y_teste) #Vai erguer uma falha e isso se deve a natureza dos modelos de classificação multiclasse
                                          #Por isso vamos usar do NumPy pra transformar esses valores e coletar um único valor por linha.
````

Agora discorreremos de maneira lenta e gradual, entendendo diversas funções dentro do TensorFlow para que então consigamos entender de maneira mais aprofundada hiperparâmetros essenciais, dentre outras especificidades que venham a surgir durante os processos de explicação. ### **Models**

É o nosso inicializador, ele é o responsável por criar toda a arquitetura da nossa rede, e é nele que definimos qual tipo de rede iremos confeccionar de acordo com a finalidade que buscamos.

````python
import tensorflow as tf
from tensorflow.keras import models, layers
#Temos três tipos de modelos que podem ser utilizados dentro da biblioteca Models.
    #São eles via API Funcional, sendo este flexível e permitindo multiplos inputs/outputs.
    #O Subclassing que é mais complexo e envolve OOP, ele traz liberdade total, contudo é hard-coded.
    #E o Sequential, que é o modelo mais simples e também lienar, conhecido por ser uma pilha de camadas.

#Agora que então entendemos os três tipos de modelos, entendemos também que a primeira etapa ao criar uma rede neural é definir sua arquitetura.
#Neste caso iremos optar pelo Sequential por sua facilidade didática, agora antes de implementarmos o modelo.
    #Precisamos entender que dentro de sua arquitetura, ou usando de analogia, dentro de um corpo temos 'n' camadas, dentre elas o esqueleto, a musculatura e a pele.
    #E dentro de sua arquitetura é a mesma coisa, o esqueleto representando a camada inicial, a musculatura os neurônios, e sua saída a pele.
````

````python
#Logo ainda aproveitando o conjunto do MNIST para dar continuidade temos o seguinte.
modelo = models.Sequential([
    layers.Input(shape = (28, 28)), #Relato que o dado de entrada possui estas dimensões.
    layers.Flatten(),                      #Achato o dado de entrada, pois os neurônios não conseguem receber nada além de linhas e colunas.
    layers.Dense(48, activation = 'tanh'), #Seleciono a quantia de neurônios e a função de ativação que irei utilizar.
    layers.Dense(12, activation = 'relu', kernel_initializer = 'he_normal'), #Criei um segundo neurônio utilizando reLU e troquei o inicializador de pesos.
                                                                             #Importante mencionar que o he_normal é um inicializador de pesos bem agressivo e que não trabalha bem com curvas suaves.

    layers.Dropout(0.6), #O Dropout tem a função de sabotar o aprendizado no intuito de evitar Overfitting, com ele desligamos parte das amostras e evitamos que sempre sejam utilizados neurônios iguais.
    layers.BatchNormalization(), #Essa camada tem como objetivo normalizar os dados permitindo que a média fique próxima a zero e o desvio perto de 1.
    layers.Dense(10, activation = 'softmax', kernel_initializer = 'glorot_uniform') #Por padrão o inicializador é o glorot, mas basicamente ele é bom com funções de ativação de curva suave
                                                                                    #O Sequential entende que por padrão o último neurônio a ser criado se torna nossa camada de saída.
    ])
````

Após criarmos o nosso modelo e suas camadas, seguimos para a compilação para definir seus otimizadores de peso, o batch, dentre outros hiperparâmetros dos quais também iremos discorrer no código através de comentários.

````python
modelo.compile(
    optimizer = 'adam', #Aqui decidimos o nosso otimizador de pesos. E se montamos um modelo via API
                        #Conseguimos modificar inclusive a velocidade da taxa de aprendizagem que impacta na sua fórmula.

    loss = 'sparse_categorical_crossentropy', #Por padrão em modelos multiclasse, mas temos para categoricos hot-encoding.
    metrics = ['accuracy'] #Podemos incluir quantas métricas precisarmos, contudo como estamos num modelo multiclasse
                           #Nem todas as métricas funcionam.

)
````

````python
modelo.fit(
    X_train,         #Primeiro hiperparâmetro sempre será o X_treino
    y_train,         #O segundo sempre será a resposta esperada ao X_treino
    epochs = 5,     #Número de vezes que o modelo irá percorrer o conjunto durante o treinamento
    batch_size = 20, #Número de amostras coletadas por rodada de treino para auxiliar os otimizadores.
                       #Quanto maior o número de batchs mais estável o resultado, porém mais custoso em RAM.
                       #Quanto menor o número de batchs, o aprendizado é mais rápido, porém oscila e perde estabilidade.

    shuffle = True, #Impede que o modelo mantenha a mesma ordem dos dados a cada epoch, por padrão = True.

    validation_split = 0.2  #Separa parte dos dados de treino para testar o modelo durante seu treinamento, o que permite
                            #identificar desde cedo se o modelo teve overfitting em relação ao material dado para treino.
)
````

### Lembram do Dropout de 60%?

Ele foi o responsável pelo modelo ter tido tão péssimo desempenho em seu treinamento e uma melhor resposta durante a validação. Acontece que removemos 60% dos neurônios durante iterações para impedir que ele utilizasse o mesmo neurônio de forma seguida.

Contudo, quando ele foi fazer a prova, pôde usar 100% dos seus neurônios e teve um desempenho muito melhor por conta disto! Logo é muito interessante o resultado, porém extremamente perigoso, logo como forma de evitar que isso ocorra, podemos também diminuir o Dropout e ver como ficaria o resultado.

Porém, antes disso iremos ver gráficamente o resultado da predição com o treinamento atual e então rodar novamente o mesmo modelo com um Dropout de 25% para entender o quão melhor o nosso modelo irá responder durante o treinamento.

````python
predito = modelo.predict(X_test)
````

````python
valor_predito = np.argmax(predito, axis = 1)
confusao = confusion_matrix(valor_predito, y_test)

sns.heatmap(confusao, annot = True, fmt = 'd', cmap = 'Blues')
plt.title('Matriz de confusão do MNIST')
plt.xlabel('Valor que o modelo previu')
plt.ylabel('Valor esperado')
plt.show()
````

````python

modelo = models.Sequential([
    layers.Input(shape = (28, 28)),
    layers.Flatten(),
    layers.Dense(48, activation = 'tanh'),
    layers.Dense(12, activation = 'relu', kernel_initializer = 'he_normal'),
    layers.Dropout(0.25),
    layers.BatchNormalization(),
    layers.Dense(10, activation = 'softmax', kernel_initializer = 'glorot_uniform')

    ])

modelo.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

modelo.fit(
    X_train,
    y_train,
    epochs = 5,
    batch_size = 20,
    shuffle = True,
    validation_split = 0.2

)
````

Perceberão o quão mais nosso modelo evoluiu durante cada época? Muito incrível!

````python
predito = modelo.predict(X_test)
````

````python
valor_predito = np.argmax(predito, axis = 1)
confusao = confusion_matrix(valor_predito, y_test)

sns.heatmap(confusao, annot = True, fmt = 'd', cmap = 'Blues')
plt.title('Matriz de confusão do MNIST')
plt.xlabel('Valor que o modelo previu')
plt.ylabel('Valor esperado')
plt.show()
````
