# Chegada das leis de proteção de dados

O Brasil foi um dos países que levaram um tempo para estabelecer uma lei que de fato nos auxiliasse no respeito à privacidade dos dados alheios, tendo a Lei Geral de Proteção de Dados Pessoais — **Lei nº 13.709/2018 (LGPD)** — promulgada apenas em 2018, entrando em vigor somente em setembro de 2020. Foi graças a esta lei que passamos a ter um olhar mais cuidadoso perante os dados advindos de pessoas individuais. Contudo, o Brasil não foi o único a instaurar leis que protegem os dados privados das pessoas.

Países como a Austrália, Canadá, Argentina, Japão, China, Estados Unidos (Califórnia) e a União Europeia instauraram leis que visam essa proteção em tempos distintos, algumas dessas leis são conhecidas como:

* **Australian Privacy Principles (APPs):** São as diretrizes estabelecidas pela Lei de Privacidade de 1988 na Austrália. Essa lei é composta por 13 princípios que regem práticas de privacidade de organizações privadas. Os APPs abordam aspectos como obtenção de consentimento, manutenção de precisão dos dados e segurança da informação pessoal.

* **Personal Information Protection and Electronic Documents Act (PIPEDA):** A PIPEDA foi promulgada nos anos 2000 no Canadá e tem como objetivo a regulação da coleta, uso e divulgação de informações pessoais por organizações privadas do setor comercial. Essa lei estabelece princípios para o tratamento ético e responsável dos dados pessoais daqueles que consentiram a coleta.

* **Protección de Los Datos Personales:** Promulgada em 30 de outubro de 2000 na Argentina (Lei nº 25.326), essa lei confere aos titulares direitos sobre seus dados pessoais, além de impor às empresas princípios para o tratamento destes dados.

* **The Act on the Protection of Personal Information (APPI):** Legislação japonesa de 2003 que estabelece diretrizes para a coleta, uso e tratamento adequado de dados pessoais por organizações. Requer consentimento para manipulação de informações, define medidas de segurança e concede aos indivíduos o direito de acessar e corrigir seus dados.

* **General Data Protection Regulation (GDPR):** Regulamento aplicável a todos os países que fazem parte da União Europeia e Espaço Econômico Europeu. Teve sua primeira publicação em 27 de abril de 2016 e **entrou em vigor em 25 de maio de 2018**. A proximidade da vigência da GDPR foi um dos fatores que acelerou a aprovação da LGPD no Brasil.

* **California Consumer Privacy Act (CCPA):** Lei destinada a aprimorar os direitos de privacidade e proteção do consumidor para residentes do estado da Califórnia, Estados Unidos. Publicada em 28 de junho de 2018.

* **Personal Information Protection Law (PIPL):** A China já garantia liberdade e privacidade em sua Constituição de 1949, mas em 2021 implementou a PIPL, que visa salvaguardar a privacidade e segurança dos dados pessoais. Aplicável a organizações locais e internacionais, estabelece diretrizes claras e categoriza alguns processadores como operadores críticos de infraestrutura.

---

A criação destas leis não garantiu apenas uma privacidade maior nos meios digitais, como também uma proteção abrangente para dados pessoais em **qualquer meio — digital ou físico**. A própria LGPD, em seu artigo 1º, explicita que dispõe sobre o tratamento de dados pessoais "**inclusive** nos meios digitais", o que significa que fichas de papel, formulários físicos, cadastros telefônicos e quaisquer outros meios também estão cobertos pela lei.

Uma das motivações centrais para a criação dessas legislações foram os abusos cometidos por organizações na utilização de dados pessoais para tomada de decisões. Empresas de saúde, por exemplo, decidiam se liberavam ou não planos de saúde com base em dados genéticos, hábitos e costumes de um cidadão. Na esfera financeira, a coleta desregulada de dados sobre a saúde financeira de indivíduos e seus familiares era utilizada para negar a concessão de crédito. Essas práticas evidenciaram a necessidade de limites legais claros para o tratamento de dados pessoais.

## **Agora que falamos sobre privacidade, como vamos proteger estes dados?**

A proteção dos dados é efetiva quando diversas camadas passam pelas proteções necessárias e têm suas devidas instruções aplicadas de maneira devida. Logo iremos dissecar cada etapa que precisa estar devidamente protegida para entender todo o processo por trás.

* **Rede externa:** A rede externa é a fronteira entre a infraestrutura da organização e a internet pública. Pense nela como o muro de um data center: tudo que vem de fora precisa passar por aqui antes de alcançar qualquer recurso interno. A maioria dos ataques se origina nesse ponto. Um atacante externo primeiro precisa transpor essa barreira para alcançar servidores, aplicações e dados internos. Se essa camada falha, todas as camadas internas ficam expostas.

* **Rede interna:**  A rede interna é tudo que está dentro do perímetro: comunicação entre estações de trabalho, servidores, bancos de dados, sistemas internos. Quando um atacante consegue ultrapassar o perímetro (ou quando a ameaça se origina internamente), é aqui que o dano real acontece. Uma vez dentro da rede, o atacante pode se mover lateralmente — ou seja, saltar de um sistema para outro buscando dados sensíveis, credenciais elevadas ou acesso a bancos de dados. Se a rede interna é "plana" (sem divisões), comprometer um único dispositivo pode significar acesso a tudo.


* **Servidores:** Servidores são as máquinas (físicas ou virtuais) que armazenam, processam e disponibilizam dados e serviços para os usuários e sistemas da organização. Aqui residem bancos de dados, aplicações corporativas, sistemas de e-mail, APIs e tudo que sustenta a operação. É aqui que os dados efetivamente residem. Um servidor comprometido pode significar acesso direto a milhões de registros pessoais, financeiros ou estratégicos. No contexto da LGPD, um vazamento nessa camada é o cenário mais grave possível.

* **Endpoint:** O endpoint é o dispositivo final onde o usuário interage com os sistemas: computadores corporativos, notebooks, smartphones, tablets. É o ponto mais próximo do fator humano e, por isso, é onde erro humano e engenharia social se manifestam diretamente. O endpoint é simultaneamente a camada mais numerosa e a mais vulnerável. Cada colaborador representa um ponto de entrada potencial. Um único clique em um e-mail de phishing pode comprometer toda a cadeia — do endpoint ao servidor, passando pela rede interna.

## **Tipos de dados**

Dentro dos tipos de dados presentes no quesito proteção de dados, temos:

* **Dados anonimizados:** Dados anonimizados são informações processadas para remover ou modificar identificadores pessoais, tornando impossível, por meios técnicos razoáveis, vincular o dado ao seu titular original. Esse processo é irreversível e, sob a LGPD (Lei Geral de Proteção de Dados), dados anonimizados não são considerados dados pessoais, liberando-os da aplicação da legislação.

* **Dado pessoal** Segundo a LGPD (Lei nº 13.709/2018), dado pessoal é toda informação relacionada a uma pessoa natural viva, identificada ou identificável. Inclui dados diretos (nome, CPF, RG, telefone, endereço) e indiretos (geolocalização, IP, hábitos de consumo, fotos) que permitam rastrear um indivíduo. A lei regula o tratamento desses dados, que deve ter finalidade clara e legítima.

* **Dado pessoal sensível:** Dados pessoais sensíveis, sob a LGPD (Lei nº 13.709/2018), são informações íntimas que, se vazadas ou usadas indevidamente, podem gerar discriminação ou graves riscos ao titular. Incluem origem racial/étnica, convicção religiosa, opinião política, filiação sindical, dados genéticos, biométricos, de saúde, exigindo tratamento especial e consentimento.

## **Tratamentos para dados pessoais**

E aqui é onde começamos de fato a trabalhar para trazer segurança aos dados particulares dos usuários que dispuseram dos seus dados e confiaram em permitir o tratamento destes dados para diversos fins de melhorias e análises.

* **Pseudonimização:** Técnica que substitui identificadores diretos por valores indiretos de forma reversível. Inclui tabelas de relacionamento (chaveamento de/para), tokenização, hashing criptográfico e mascaramento estruturado. Dados pseudonimizados ainda são considerados dados pessoais.

* **Criptografia:** Transformação reversível de dados em formato ilegível mediante chave criptográfica, protegendo confidencialidade durante armazenamento e transmissão. É medida de segurança técnica. Dados criptografados permanecem sendo dados pessoais.

* **Gestão de acessos:** Permitir o acesso apenas para pessoas autorizadas, impedindo que quaisquer outros não tenham acesso expresso aos arquivos sejam estes físicos ou digitais.

* **Auditoria de logs:**  Registro sistemático e análise de eventos de acesso e
modificação de dados. Permite rastreabilidade completa (quem acessou o quê,
quando e de onde), detecção de acessos não autorizados e investigação de
incidentes. Logs devem ser imutáveis, com timestamp confiável e armazenamento
separado dos sistemas auditados.

* **Backups** Backups programados para evitar perda de dados, e perguntas como "aquele dado de fato foi perdido para sempre?" isso também faz parte dos processos para garantir a segurança dos dados coletados e tratados, sendo que quando os dados perdem sua usabilidade ou o usuário solicita a retirada dos dados, os backups dos dados destes devem ser removidos também.

Após entendermos um pouco sobre os processos de tratamentos para os dados pessoais, seguimos com os princípios de livre acesso, qualidade dos dados e transparência que refletem diretamente ao fato de ser direito do usuário o acesso aos seus dados quando dados pessoais/dados pessoais sensíveis, e que para isso é necessário entender que os dados coletados sempre terão um processo linear e estruturado de utilização, onde temos a coleta, tratamento e exclusão dos dados de acordo com diretrizes de tempo pré-estabelecidas ou a pedido do usuário que naquele momento permitiu a coleta dos seus dados e em outro momento solicitou a exclusão. Salvo alguns casos em que o tratamento dos dados tramita com finalidade de execuções contratuais dentre outros exemplos.

**Sempre é interessante trazer dicas para o entendimento pleno, então temos o Art. 7º da Lei nº13.709 (LGPD) que nos traz informações das bases legais para o tratamento de dados pessoais. ## **Cookies & Leads**

Apesar de Cookies terem o significado literal de biscoitos, no mundo da LGPD os Cookies são pequenos arquivos que ficam armazenados em navegadores e que desempenham papel fundamental na coleta de informações sobre o comportamento de um usuário dentro de um determinado site, **além disto o usuário que têm de possuir o direito de negar os cookies dependendo da finalidade do site, assim como aceitar somente alguns ou todos os cookies presentes na página.** Por fim, essas configurações precisam estar presentes de forma explicita de forma que o usuário consiga aceitar, negar ou aceitar de forma granular a coleta dos dados durante a sua navegação.

Os Cookies podem ter diversas finalidades, como por exemplo para análises estatísticas, publicidade e remarketing, análise de comportamento, dentre algumas outras finalidades. É importante também mencionar que estes podem ser temporários (expiram após encerramento da sessão), persistentes (permanecem durante toda a jornada do usuário até que se expirem de forma programada ou sejam excluidos pelo usuário), além disto os Cookies dentro de um site podem vir de sites de terceiros ou do próprio site acessado.

**OBS:** Existem os Cookies essenciais que são estritamente necessários para o funcionamento do site, logo não podem ser rejeitados pelo usuário.

---

**Lead** é um contato qualificado que demonstrou interesse em produtos, serviços ou conteúdos de uma empresa através de alguma ação voluntária, fornecendo informações de contato que permitem sua identificação e comunicação futura. Para que um usuário anonimo se torne um **Lead** é necessário que durante o tratamento dos dados deste, hajam meios de contato direto, geralmente sendo o e-mail indispensável para qualificação de um Lead.
* Empresa
* Telefone
* Nome
* E-mail

Que haja a manifestação do interesse sobre o produto de alguma forma, seja por participação de uma webinar, preenchimento de formulário, inscrição em newsletter ou solicitações de demonstrações ou orçamentos. Além da permissão para contato que deve ser específico e informado ao usuário de forma que o mesmo conceda a permissão durante seu acesso a quaisquer que sejam os eventos.

Por fim quando um visitante anônimo rastreado por cookies se torna um lead ao fornecer seus dados, os sistemas integram as informações comportamentais anteriormente coletadas (páginas visitadas, origem do tráfego, tempo de navegação) ao registro identificável do lead, criando um perfil completo que permite comunicação personalizada baseada no histórico de interesse demonstrado.

**CUIDADO:** Os cookies podem ser categorizados como dados anonimizados, contudo quando há a qualificação de Lead há possibilidade do cruzamento e identificação dos dados. Logo, quando isto ocorre é necessário garantir os direitos sobre os dados que aquele usuário terá, além de tratar toda sua jornada antes anonimizada como agora um dado pessoal identificado.

## **Rolou vazamento, e agora?**

A LGPD traz que quando há incidentes de segurança que possam causar riscos ou danos relevantes aos dados dos usuários, que a empresa comunique tanto à Autoridade Nacional quanto os titulares que podem ser afetados incluindo informações essenciais de quais dados foram afetados, quais riscos estão associados, quais foram as medidas para proteger os dados e o motivo pela demora na comunicação, além das medidas tomadas para reverter ou mitigar estes prejuizos.

Agora quando de fato há o vazamento a LGPD estabelece responsabilidades e ressarcimento de danos aos usuários afetado, além da empresa passar por sanções administrativas pela Autoridade Nacional

**OBS:** Neste caso a Autoridade Nacional é a Agência Nacional de Proteção de Dados (ANPD).

## **Tipos de consentimento**

O consentimento do titular precisa ser livre, informado, inequívoco e utilizado para uma finalidade determinada. Ou seja:
* **Livre:** O titular precisa ser capaz de tomar uma decisão real e autônoma sem nenhum tipo de pressão, consequências negativas pelo não consentimento ou coerção. Contudo, nos casos de desequilíbrio de poder, como entre o poder público e o titular o consentimento pode não ser válido, apesar de existirem exceções.

* **Informado:** O titular precisa ser informado sobre tudo que será utilizado acerca dos seus dados antes de consentir com quaisquer termos. Tendo informações como por exemplo; finalidade, tratamento, tipos de dados a serem coletados. Além de ter o direito de retirar o consentimento e ter ciência dos possíveis riscos de consentir com o tratamento dos dados.

* **Inequívoco:** O titular precisa consentir de forma clara e afirmativa sua autorização'

## LGPD Breve

A LGPD (Lei nº 13.709/2018) visa proteger direitos fundamentais de
liberdade, privacidade e livre desenvolvimento da personalidade de
cada indivíduo, regulando o tratamento de dados pessoais realizado
por pessoas físicas ou jurídicas, de direito público ou privado,
em meios físicos ou digitais.

A LGPD se aplica quando:
• A operação de tratamento ocorre no Brasil, OU
• Há oferta de bens/serviços ao mercado brasileiro, OU
• Dados de pessoas localizadas no Brasil são tratados

Isso significa que empresas estrangeiras também estão sujeitas à LGPD
se oferecerem produtos/serviços ao público brasileiro, independentemente
da localização de seus servidores ou da nacionalidade dos titulares.

Já a LGPD não se aplica para:

1. **Fins exclusivamente jornalísticos, artísticos ou acadêmicos**
   (não abrange atividades comerciais/administrativas dessas organizações)

2. **Segurança pública, defesa nacional, segurança do Estado**

3. **Investigação e repressão de infrações penais**

4. **Pessoa natural para fins exclusivamente particulares e não econômicos**
   • Exemplos: agenda pessoal de contatos, álbum de fotos familiar
   • Não se aplica se houver finalidade econômica (ex.: freelancer
     gerenciando base de clientes)

5. **Tratamento fora do Brasil** sem comunicação, compartilhamento com
   agentes brasileiros ou transferência internacional envolvendo o Brasil
